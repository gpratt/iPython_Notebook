```python
fdr_value = .05
fdr_found = False
x_axis = []
y_axes = defaultdict(list)
for cur_cutoff in np.arange(min(combined_dfs.ix["FLAG"].log2Ratio), 
                         max(combined_dfs.ix["FLAG"].log2Ratio), .01):
    cur_result = combined_dfs[combined_dfs.log2Ratio > cur_cutoff]
    fdr = len(cur_result.ix["FLAG"].dropna()) / float(len(combined_dfs.ix["FLAG"].dropna()))
    if fdr < fdr_value and not fdr_found:
        cutoff = cur_cutoff
        fdr_found = True
    x_axis.append(fdr)
    for x in set(combined_dfs.index.get_level_values(0)):
        y_axes[x].append(len(cur_result.ix[x].dropna()))
x_axis = np.array(x_axis)
y_axes = {name : np.array(value) for name, value in y_axes.items()} 
```

```python
def loglog_kde(series1, series2, ax=None, **kwargs):
    foo = pd.concat([series1, series2], axis=1, join="inner")
    foo = np.log10(foo)
    foo[np.isneginf(foo) | np.isinf(foo)] = np.nan
    foo = foo.dropna()

    if ax:
        return sns.kdeplot(foo[foo.columns[0]], foo[foo.columns[1]], ax=ax, **kwargs)
    else: #not as ax:                                                                                                                          
        return hist2d(foo.one, foo.two, bins=175, norm=mpl.colors.Normalize(), cmin=1, vmin=0, vmax=45 ,**kwargs)
```

```python
#loads all peaks irrespective of p-value
def get_unfiltered_peaks(loaded_peaks):
    peaks = []
    for gene in loaded_peaks:
        try:
            peaks += [pybedtools.create_interval_from_list([peak.chrom, 
                                                        str(peak.genomic_start), 
                                                        str(peak.genomic_stop), 
                                                        peak.gene_name, 
                                                        str(max(peak.gene_poisson_p, peak.super_local_poisson_p)), 
                                                        peak.strand]) 
                  for peak in gene['clusters']]
        except TypeError:
            print gene
            raise
    return pybedtools.BedTool(peaks).sort(stream=True)
```

```python
def get_first_regions(interval, region_length=200):
    region_length = min(region_length, len(interval) - 1)
    if interval.strand == "+":
        interval.stop = interval.start + region_length
    else:
        interval.start = interval.stop - region_length
    return interval

def get_remaining_regions(interval, region_length=200):
    region_length = min(region_length, len(interval) - 1)
    if interval.strand == "+":
        interval.start = interval.start + region_length
    else:
        interval.stop = interval.stop - region_length
    return interval
```

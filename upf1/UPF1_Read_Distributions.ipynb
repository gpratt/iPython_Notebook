#Analysis Logic

```python
def get_bam_coverage(bamfile):
    bam = HTSeq.BAM_Reader(bamfile)
    coverage = HTSeq.GenomicArray("auto", typecode="i", stranded=True)
    for read in bam:
        if read.aligned:
            for cigop in read.cigar:
                if cigop.type != "M":
                    continue
                coverage[cigop.ref_iv] += 1
    return coverage

def get_bam_counts(bamfile):
    bam = HTSeq.BAM_Reader(bamfile)
    coverage = HTSeq.GenomicArray("auto", typecode="i")
    for read in bam:
        read.iv.length = 1
        if read.aligned:
            coverage[read.iv] += 1
    return coverage

def bed_to_genomic_interval(bed):
    for interval in bed:
        yield HTSeq.GenomicInterval(interval.chrom, interval.start, interval.stop, interval.strand)

def adjust_width(interval):
    width = 250
    interval.start -= width
    interval.stop += width
    return interval

def is_protein_coding(interval):
    return interval.name in protein_coding_genes.index

def gff_to_bed(gff_handle):
    for interval in gff_handle:
        yield pybedtools.create_interval_from_list([interval.chrom, 
                                                 str(interval.start),
                                                 str(interval.stop),
                                                 interval.attributes['gene_id'][0], 
                                                 "0",
                                                 interval.strand])

def get_densities(intervals, coverage):
    for interval in bed_to_genomic_interval(intervals):
        density = numpy.fromiter(coverage[interval], dtype="i")
        if interval.strand == "-":
            density = density[::-1]
        yield density
        
def get_coverage_around_features(features):
    """
    
    Given a bedtool of features retrurns for UPF1 read coverage around those features
    
    """
    
    wt_coverages = []
    deaa_coverages = []
    k498a_coverages = []
    features = features.filter(is_protein_coding).each(adjust_width).saveas()
    names = [feature.name for feature in features]
    df_wt = pd.DataFrame(list(get_densities(features, wt_coverage)), 
                         index=names).fillna(0)
    df_deaa = pd.DataFrame(list(get_densities(features, deaa_coverage)),
                           index=names).fillna(0)
    df_k498a = pd.DataFrame(list(get_densities(features, k498a_coverage)),
                            index=names).fillna(0)
    
    df = pd.concat({"wt" : df_wt, "deaa" : df_deaa, "k498a" : df_k498a})
    return df

def coding_transcript_poly_a_sites(db):
    transcripts = db.features_of_type("transcript")
    for x, transcript in enumerate(transcripts):
        if (transcript.attributes['gene_type'][0] == "protein_coding"
            and transcript.attributes['transcript_type'][0] == "protein_coding"):
            if transcript.strand == "+":
                transcript.start = transcript.stop
                transcript.stop = transcript.stop + 1
            else: #negative strand
                transcript.stop = transcript.start
                transcript.start = transcript.start - 1
            yield transcript
            
def coding_transcription_start_sites(db):
    transcripts = db.features_of_type("transcript")
    for x, transcript in enumerate(transcripts):
        if (transcript.attributes['gene_type'][0] == "protein_coding"
            and transcript.attributes['transcript_type'][0] == "protein_coding"):
            if transcript.strand == "-":
                transcript.start = transcript.stop
                transcript.stop = transcript.stop + 1
            else: #positive strand
                transcript.stop = transcript.start
                transcript.start = transcript.start - 1
            yield transcript
```

```python
def bin_evenly(lst, bin_size):
    """
    
    Bins into evenly dispersed groups, only shrinks things, won't work for things smaller than bin_size

    """
    
    if type(lst) is float: #nan check
        empty = np.empty(100)[:]
        empty[:] = np.nan
        return empty
    
    empty = np.zeros(bin_size - (len(lst) % bin_size))
    empty[:] = np.nan
    binned_results = np.mean(np.concatenate([lst, empty]).reshape((bin_size,-1)), axis=1)
    return binned_results
```

```python
def get_reads_around_introns(gene_dicts):
    """
    
    Takes dict of lists of genes of read distributions and outputs read denisites around introns
    
    """
    densities = []
    for gene in gene_dicts.values():
        for intron in range(len(gene) -1):
            #introns aren't always going to be 50bp wide so pad them with zeros if they aren't
            left_side = zeros(50)
            left_side[-len(gene[intron][-50:]):] = gene[intron][-50:]
            right_side = zeros(50)
            right_side[:len(gene[intron + 1][:50])] = gene[intron + 1][:50]
            if len(left_side) != 50:
                print "left side broken", len(left_side)
                print left_side
                raise
            densities.append(np.concatenate([left_side, right_side]))
    
    for gene in densities:
            gene[np.isnan(gene)] = 0
            
    densities = array([list(abs(density)) for density in densities if np.mean(density) > 2])
    return densities
```

```python
def get_densitites_by_region(coverage):
    """
    
    Given a coverage vector returns a densities across combined 5' UTRs, CDS and 3' UTRs for all genes
    as a pandas dataframe. 
    
    """
    genes = set(actual_distributions['utr5'].keys() + 
    actual_distributions['cds'].keys() + 
    actual_distributions['utr3'].keys())

    densities = defaultdict(dict)
    for gene in genes:
        regions = {}
        for region in ['utr5', 'cds', 'utr3']:
            regions[region] = {x : density for x, density in  enumerate(get_densities(actual_distributions[region][gene], 
                                                          coverage))}
        densities[gene] = pd.DataFrame(regions)

    densities = pd.concat(densities)
    return densities

def get_counts_by_region(coverage):
    """
    
    Given a coverage vector returns a densities across combined 5' UTRs, CDS and 3' UTRs for all genes
    as a pandas dataframe. 
    
    Compresses to counts and length quickly to avoid out of memory errors (happens when dealing with introns)
    """
    
    genes = set(actual_distributions['utr5'].keys() + 
    actual_distributions['cds'].keys() + 
    actual_distributions['utr3'].keys())

    densities = defaultdict(dict)
    for gene in genes:
        regions = defaultdict(dict)
        for region in ['utr5', 'cds', "introns", 'utr3']:
             densities[gene][region] = [{"count" : sum(item), "length" : len(item)} for item in get_densities(actual_distributions[region][gene], 
                                                                                                         coverage)]
 
    densities = pd.DataFrame(densities).T
    return densities
```

```python
def robust_concatenate(lst):
    try:
        return np.concatenate(lst)
    except:
        return np.nan
    
intermediate_density_df = defaultdict(dict)
for index, series in density_frame.groupby(level=[0,1]):
    experiment, gene = index
    regions = {}
    for region in series.columns:
        regions[region] = robust_concatenate(series[region].dropna())
    
    intermediate_density_df[experiment][gene] = pd.Series(regions)
    
tmp = {}
for experiment in intermediate_density_df.keys():
    tmp[experiment] = pd.concat(intermediate_density_df[experiment])
intermediate_density_df = pd.concat(tmp).unstack()
del tmp
```

```python
def normalize_plot(series):
    return series / series.sum()
```

```python
def significant_bases(df):
        number_of_bases = len(df.columns)
        mean = df.sum(axis=1).mean() / number_of_bases
        means = df.mean()
        
        z = (means - mean) / (df.std() / np.sqrt(len(df)))
        return np.array((stats.norm.pdf(z) < (.05 / number_of_bases)) & (means > mean))
```

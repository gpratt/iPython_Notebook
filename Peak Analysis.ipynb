```python
import pybedtools
from bx.bbi.bigwig_file import BigWigFile
from sklearn.decomposition import PCA
```

```python
def get_peak_wiggle(tool, data, size = 100, num_peaks=100):
    
    """
    
    returns matrix of (n,25) of wiggle and peak information
    tool - bedtool contains location of peaks
    data - bigwig file handle - contains bw data for the peaks

    """
    
    results = []

    for line in tool[:num_peaks]:
        peak_len = (line.stop - line.start)
        if peak_len > size:
            #get rid of this when we parse down larger peaks
            continue
            
        if  peak_len < size:
            front_amount = (size - peak_len) / 2 + (peak_len % 2)
            back_amount = (size - peak_len) / 2 

        result = data.get_as_array(line.chrom, line.start - front_amount , line.stop + back_amount)
        result = [0 if np.isnan(x) else x for x in result]
        
        if len(result) != size:
            print "something strange"
            continue
            
        results.append(result)
        #print len(result)

    return array(results)
```

```python
def get_coordinates(df, selectors):
    """
    
    Given a df with only cyto, insol, memb and nuc indexes converts to polar coordinates
    
    """
    
    angles = np.arange(0, 2*np.pi, 2*np.pi / len(selectors))
    df = df.ix[selectors]
    
    percent_df = (df / df.sum()).dropna(axis=1)
    
    x_cor = pd.Series(index=percent_df.columns)
    y_cor = pd.Series(index=percent_df.columns)
    x_cor[:] = 0
    y_cor[:] = 0
    
    for selector, angle in zip(selectors, angles):
        x_cor += percent_df.ix[selector].apply(lambda x: x * np.cos(angle))
        y_cor += percent_df.ix[selector].apply(lambda x: x * np.sin(angle))

    return pd.concat({"x" : x_cor, "y" : y_cor}).unstack()
```

```python
OFFSET = .20
def simplex_graph(coordinates, genes, selectors, ax, bound_name=""):
    ax.scatter(coordinates[list(set(coordinates.columns) - genes)].ix['x'], 
               coordinates[list(set(coordinates.columns) - genes)].ix['y'], 
               alpha=.05,
               color="0.50",
               label="Not Bound",
               )
        
    ax.scatter(coordinates[list(coordinates.columns & genes)].ix['x'], 
               coordinates[list(coordinates.columns & genes)].ix['y'], 
               alpha=.05,
               color="b",
               label="Bound",
               )
    ax.set_xticks([])
    ax.set_yticks([])
    
    angles = np.arange(0, 2*np.pi, 2*np.pi / len(selectors))
    for selector, angle in zip(selectors, angles):
        x, y = (np.cos(angle), np.sin(angle))
        
        if selector == "Membrane":
            y += .1
        ax.annotate(selector, xy=(x,y))
            
    ax.legend([mpl.lines.Line2D([0],[0], linestyle="None", marker="o", color='b'),
                mpl.lines.Line2D([0],[0], linestyle="None", marker="o", color='0.50'),],
              [bound_name + " Bound", bound_name + " Not Bound"], 
              loc=2,
              fontsize=10,
              numpoints=1,
              bbox_to_anchor=(.8, 1)
              )
    
    sns.despine(ax=ax, left=True, bottom=True)

    return ax
```

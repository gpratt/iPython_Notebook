```python
import os
import glob
from collections import defaultdict
from itertools import groupby
import multiprocessing
import pandas as pd
import pysam
from IPython.core.display import HTML

from gscripts.clipseq import cross_contamination_detector
```

```python
%autoreload 2
```

```python
pool = multiprocessing.Pool(multiprocessing.cpu_count())
```

```python
bams = glob.glob("/nas3/gpratt/projects/encode/v7/bams/*.bam")
bam_groups = groupby(sorted(bams, key=lambda x: x.split(".")[1].split("_")[0]), lambda x: x.split(".")[1].split("_")[0])

groups = {}
for group, items in bam_groups:
    groups[group] =  list(items)

tasks = []
matched_count_dict = defaultdict(dict)
total_count_dict = defaultdict(dict)
for group, items in groups.items():
    matched_count_dict[group] = defaultdict(dict)
    total_count_dict[group] = defaultdict(dict)
    for item1 in items:
        for item2 in items:
            name1 = os.path.basename(".".join(item1.split(".")[:2]))
            name2 = os.path.basename(".".join(item2.split(".")[:2]))
            out_file = os.path.join("/nas3/gpratt/projects/encode/v7/bams/out/", name1 + "_" + name2 + ".sam")
            tasks.append((item1, item2, out_file))
```

```python
results = []
jobs = [pool.apply_async(cross_contamination_detector.correlation, job) for job in tasks]
for job in jobs:
    name1, name2, matched_count, total_count, out_file = job.get()
    with open(os.path.splitext(out_file)[0] + ".txt", 'w') as result_file:
        result_file.write("\t".join(map(str, [name1, name2, matched_count, total_count])) + "\n")
```

```python
figsize(6,20)


jaccard_frame = matched_count_r01_df / total_count_r01_df.astype(float)
y_events = scipy.spatial.distance.pdist(jaccard_frame)
Z_events = scipy.cluster.hierarchy.linkage(y_events)
d_events = scipy.cluster.hierarchy.dendrogram(Z_events, orientation='right', distance_sort='descending', no_labels=True)
event_order = d_events['leaves'][::-1]
title("clustering of samples by jacard correlation") 
```

```python
import pandas as pd

img_dir = "/nas3/gpratt/Dropbox/encode_integration/figures/"
import os
import numpy as np

import seaborn as sns
from matplotlib_venn import venn2, venn3
from gscripts.general import dataviz
from scipy.stats import hypergeom

import matplotlib as mpl
from matplotlib import rc

mpl.rcParams['svg.fonttype'] = 'none'

rc('text', usetex=False)
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
%matplotlib inline
```

#Load FRIP and bound peak information

```python
exintrons = pd.read_table("/nas3/gpratt/projects/encode_integration/data/frip/exintrons_gene_exp.diff", index_col=0)
exons = pd.read_table("/nas3/gpratt/projects/encode_integration/data/frip/exons_gene_exp.diff", index_col=0)
exons.test_stat = exons.test_stat * -1
exintrons.test_stat = exintrons.test_stat * -1
imp1_a = pybedtools.BedTool("/nas3/gpratt/projects/encode/analysis/meeting/IMP1_A.peaks.bed")
imp1_c = pybedtools.BedTool("/nas3/gpratt/projects/encode/analysis/meeting/IMP1_C.peaks.bed")
imp1_a_names = set((interval.name.split("_")[0] for interval in imp1_a))
imp1_c_names = set((interval.name.split("_")[0] for interval in imp1_c))
```

```python
sig_exons = exons[exons.significant == "yes"]
sig_exintrons = exintrons[exintrons.significant == "yes"]
```

#How well do FRIP significant peaks overlap with things?

```python
venn2([set(sig_exons.gene_id.values), set(sig_exintrons.gene_id.values)], set_labels=["exons", "exintrons"])
```

```python
with dataviz.Figure(os.path.join(img_dir, "frip_overlapping_genes.svg"), figsize=(3,3)) as fig:
    ax = fig.add_subplot(1,1,1)
    venn2([set(sig_exons[sig_exons.test_stat > 0].gene_id.values), imp1_a_names], 
      set_labels=['IMP1 FRIP', 'IMP1 CLIP A'],
      ax=ax)
    ax.set_title("FRIP and CLIP\nOverlapping Targets")
```

```python
with dataviz.Figure(os.path.join(img_dir, "frip_overlapping_genes.svg"), figsize=(6,6)) as fig:
    ax = fig.add_subplot(1,1,1)
    venn3([set(sig_exons.gene_id.values), imp1_a_names, imp1_c_names], 
      set_labels=['IMP1 FRIP', 'IMP1 CLIP A', 'IMP1 CLIP B'],
      ax=ax)
```

#More importantly are the score distributions different between bound and unbound sets?

```python
print len(sig_exons[sig_exons.gene_id.apply(lambda x: x in imp1_a_names)])
print len(sig_exons[sig_exons.gene_id.apply(lambda x: x not in imp1_a_names)])
```

```python
with dataviz.Figure(os.path.join(img_dir, "frip_enrichment.svg"), figsize=(3,3)) as fig:
    ax = fig.add_subplot(1,1,1)
    dataviz.plot_cdf(sig_exons[sig_exons.gene_id.apply(lambda x: x in imp1_a_names)].test_stat, ax=ax, label="Bound in IMP1 CLIP")
    dataviz.plot_cdf(sig_exons[sig_exons.gene_id.apply(lambda x: x not in imp1_a_names)].test_stat, ax=ax, label="Unbound in IMP1 CLIP")
    ax.legend(loc=0, bbox_to_anchor=(1.0, 1))
    ax.set_title("FRIP Scores\nSplit by Binding")
    ax.set_ylabel("Cumulative Fraction")
    ax.set_xlabel("IMP1 FRIP Score")
    sns.despine(fig)
    
```

```python
import pandas as pd
from itertools import groupby
from IPython.core.display import HTML
```

```python
metrics_files = !ls *.metrics
```

```python
metrics_dfs = {metrics_file.split(".")[1] + "_" + ".".join(metrics_file.split(".")[7:9]) : pd.read_csv(metrics_file, sep="\t", index_col=0) for metrics_file in metrics_files}
```

```python
total_counts = {}
remaining_counts = {}
for name, group in groupby(sorted(metrics_dfs.keys()), key=lambda x: x[:-4]):
    total_counts[name] = {}
    remaining_counts[name] = {}
    for item in group:
        loc = item[-3:]
        sumed_metrics = metrics_dfs[item].sum()
        remaining_count = sumed_metrics['total_count'] - sumed_metrics['removed_count']
        total_counts[name][loc] = sumed_metrics['total_count']
        remaining_counts[name][loc] = remaining_count
total_counts = pd.DataFrame(total_counts)
remaining_counts = pd.DataFrame(remaining_counts)
```

```python
HTML(remaining_counts.to_html())
```

```python
HTML(total_counts.to_html())
```

```python
downsampling_rate = (remaining_counts / remaining_counts.ix['1.0'].astype(float))
```

```python
figsize(10,10)
for x in downsampling_rate:
    plot(arange(.1,1.1, .1), downsampling_rate[x], label=x)
title("% Downsample vs % reads remaining compared to total removed")
#legend(loc="best")
```

So this answers the question can we downsample with impunity, the answer is no.  Downsampling is almost linear, fewer total reads leads to fewer usable reads.  

Next question is: How many reads to we need sequenced to get to the 2.5M required read depth that I want

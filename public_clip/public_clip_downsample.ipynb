```python
import pybedtools

from itertools import groupby
from collections import defaultdict

import pandas as pd
import scipy.stats
import pysam

from IPython.core.display import HTML
```

```python
cd /nas3/gpratt/projects/public_clip/downsample/
```

```python
beds = !ls *.bed
```

```python
bed_lens = {}
for bed in beds:
     bed_len = !wc -l $bed
     bed_lens[bed] = int(bed_len[0].split()[0])
   
```

```python
sizes = defaultdict(list)
for name, count in bed_lens.items():
    sizes[".".join(name.split(".")[7:9])].append(count)
```

```python
peak_counts  = defaultdict(list)
read_counts  = defaultdict(list)
index = []
groups = groupby(beds, key=lambda x: x.split(".")[0])
for name, group in groups:
    group = list(group)
    if len(group) == 10:
        index.append(name)
        for item in group:
            peak_counts[".".join(item.split(".")[7:9])].append(bed_lens[item])
            read_counts[".".join(item.split(".")[7:9])].append(pysam.Samfile(".".join(item.split(".")[:-2] + ["bam"])).mapped)
sizes_df = pd.DataFrame(peak_counts)
reads_df = pd.DataFrame(read_counts)
```

```python
def rsquared(x, y):
    """ Return R^2 where x and y are array-like."""

    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    return r_value**2
```

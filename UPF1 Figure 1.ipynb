```python
def generate_distribution(dist):
    count, bins = np.histogram(dist, range=(0, 1.0), 
                                    bins=50, 
                                    density=True
                                    )
    return count, [(bins[x] + bins[x+1]) / 2 for x in range(len(bins) - 1)]
```

```python
figsize(10,10)
k498a_count, bins = generate_distribution(pickled_files['K498A']['mRNA_positions'])
DEAA_count, bins = generate_distribution(pickled_files['DEAA']['mRNA_positions'])
WT_count, bins = generate_distribution(pickled_files['WT']['mRNA_positions'])

plot(bins, k498a_count, label="K498A")
plot(bins, DEAA_count, label="DEAA")
plot(bins, WT_count, label="WT")
title("Distribution of Peaks across mRNA")
legend(loc=0)
```

```python
#probably should quantify how much data I'm loosing...
def plot_distance_from_region_combined(closest_bed):
    """ wraps intersect bed, where region of interest is the thing to plot the distance from, and bedtool is the distance 
        returns closest bed object, plots everything on one plot
        This is useful to add to clip_analysis
    """
    
    for name, closest in sorted(closest_bed.items()):
        full_data = array([ int(result[-1]) for result in closest])
        y, x =  numpy.histogram(full_data, range=(-200, 200), bins=25, normed=True)
        x = [(x[n] + x[n+1]) / 2 for n in range(len(x) - 1)]
        plot(x,y, label=name, alpha=.7)
    legend(loc=0)
```

```python
figsize(20,10)
subplot(1,2,1)
plot_distance_from_region_combined({name : bedtool for name, bedtool in transcription_stop_closest.items() if "shuffled" not in name})
#ylim(0,.005)

subplot(1,2,2)
plot_distance_from_region_combined({name : bedtool for name, bedtool in transcription_stop_closest.items() if "shuffled" in name})
#ylim(0,.005)
```

```python
WT_shuffled = []
for name, item in [(name, item) for name, item in transcription_stop_closest.items() if "WT_shuffled" in name]:
    full_data = array([ int(result[-1]) for result in item])
    WT_shuff_y, WT_shuff_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)
    WT_shuffled.append(WT_shuff_y)
    
DEAA_shuffled = []
for name, item in [(name, item) for name, item in transcription_stop_closest.items() if "DEAA_shuffled" in name]:
    full_data = array([ int(result[-1]) for result in item])
    DEAA_shuff_y, WT_shuff_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)
    DEAA_shuffled.append(DEAA_shuff_y)
    
K498A_shuffled = []
for name, item in [(name, item) for name, item in transcription_stop_closest.items() if "K498A_shuffled" in name]:
    full_data = array([ int(result[-1]) for result in item])
    K498A_shuff_y, WT_shuff_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)
    K498A_shuffled.append(K498A_shuff_y)
 
full_data = array([ int(result[-1]) for result in transcription_stop_closest['WT']])
WT_y, WT_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)

full_data = array([ int(result[-1]) for result in transcription_stop_closest['DEAA']])
DEAA_y, DEAA_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)

full_data = array([ int(result[-1]) for result in transcription_stop_closest['K498A']])
K498A_y, K498A_x = numpy.histogram(full_data, range=(-200, 10), bins=25, normed=True)
    
```

```python
WT_combined = WT_shuffled[0]
for x in range(1, len(WT_shuffled)):
    WT_combined += WT_shuffled[x]
    
WT_combined = WT_combined / len(WT_shuffled)

DEAA_combined = DEAA_shuffled[0]
for x in range(1, len(DEAA_shuffled)):
    DEAA_combined += DEAA_shuffled[x]
    
DEAA_combined = DEAA_combined / len(DEAA_shuffled)

K498A_combined = K498A_shuffled[0]
for x in range(1, len(K498A_shuffled)):
    K498A_combined += K498A_shuffled[x]
    
K498A_combined = K498A_combined / len(K498A_shuffled)
```

```python
WT_sn = WT_y / WT_combined
DEAA_sn = DEAA_y / DEAA_combined
K498A_sn = K498A_y / K498A_combined
x = [(K498A_x[n] + K498A_x[n+1]) / 2 for n in range(len(K498A_x) - 1)]
```

```python
figsize(10,10)
plot(x, WT_y, label="wt")
plot(x, DEAA_y, label="deaa")
plot(x, K498A_y, label="k498")

plot(x, WT_combined, label="wt_shuffled")
plot(x, DEAA_combined, label="deaa_shuffled")
plot(x, K498A_combined, label="k498_shuffled")
legend(loc=0)
```

```python
figsize(10,10)
plot(x, WT_sn, label="wt")
plot(x, DEAA_sn, label="deaa")
plot(x, K498A_sn, label="k498")
legend(loc=0)
```

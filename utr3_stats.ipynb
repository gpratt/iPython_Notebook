```python
from collections import defaultdict

import pandas as pd 

from clipper.src import get_genomic_regions
import gffutils
```

```python
db = gffutils.FeatureDB("/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db")
genomic_features = get_genomic_regions.GenomicFeatures("hg19", db)
genomic_regions = genomic_features.get_genomic_regions()
three_prime_utrs = genomic_regions['three_prime_utrs']
```

```python
#Get combined 3' UTR lengths
three_prime_utrs_len = defaultdict(int)
for interval in three_prime_utrs:
    three_prime_utrs_len[interval.name] += len(interval)
    
three_prime_utrs_len = pd.Series(three_prime_utrs_len)
```

```python
result = defaultdict(dict)
db = gffutils.FeatureDB("/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db")
for feature in db.features_of_type(featuretype="transcript"):
    utr5, utr3 = genomic_features._get_utrs(feature, list(db.children(feature, featuretype="CDS")), feature_types)
    #if utr5 != [] and utr3 != []:
    result[feature.attributes['gene_id'][0]][feature.attributes['transcript_id'][0]] = {'utr3' : utr3, 'utr5' : utr5}
```

```python
result_lens = defaultdict(dict)

for gene in result:
    for transcript in result[gene]:
        result_lens[gene][transcript] = {key : [len(item) for item in values] for key, values in result[gene][transcript].items()}
```

```python
utr3_lens = defaultdict(dict)

for gene in result:
    for transcript in result[gene]:
        utr3_lens[gene][transcript] = result_lens[gene][transcript]['utr3'] 
```

```python
#get total length of all 3' UTRs per transcript
utr3_total_lens = defaultdict(dict)

for gene in result:
    for transcript in result[gene]:
        utr3_total_lens[gene][transcript] = sum(utr3_lens[gene][transcript])
```

```python
#combine results into a list per gene
utr3_combined_total_lens = defaultdict(dict)

for gene in result:
    for transcript in result[gene]:
        utr3_combined_total_lens[gene] = utr3_total_lens[gene].values()
```

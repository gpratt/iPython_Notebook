```python
riboseq_df = pd.read_table("/nas3/gpratt/projects/imp_family/analysis/imp1_riboseq_v1/all_counts.txt", comment="#", header=1, index_col=0)
riboseq_count_df = riboseq_df.drop([u'Chr', u'Start', u'End', u'Strand', "Length"], axis=1)

riboseq_count_df = riboseq_count_df.rename(columns={"Control_shRNA_1_ribo.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "Control_1",
                                                    "Control_shRNA_2_ribo_GCCAAT_L005_R1.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "Control_2",
                                                    "IMP1_shRNA_1_ribo.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "IMP1_1",
                                                    "lane4_Undetermined_L004_R1.IMP1_2_ribo.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "IMP1_2",
                                                    "lane5_Undetermined_L005_R1.IMP2_1_ribo.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "IMP2_1",
                                                    "lane6_Undetermined_L006_R1.IMP2_2_ribo.polyATrim.adapterTrim.rmRep.sorted.rg.bam" : "IMP2_2",
                                                    })
riboseq_count_df = riboseq_count_df.rename(columns=lambda x: x.split(".")[0])

riboseq_rpkm_df = (riboseq_count_df * 1000000) / riboseq_count_df.sum()
riboseq_rpkm_df = riboseq_rpkm_df.div((riboseq_df.Length / 1000), axis=0)
```

```python
combined_df = pd.concat({"riboseq" : riboseq_rpkm_df.unstack(), "rnaseq" : rnaseq_rpkm_df.unstack()}, axis=1)

new_index = []
for index in combined_df.index:
   cell_type, gene = index
   new_index.append((cell_type.split("_")[0], cell_type.split("_")[1], gene))
combined_df.index = pd.MultiIndex.from_tuples(new_index, names=["cell_type", "rep", "gene_id"])
```

```python
def geneNames(x):
    try:
        return gene_name_dict[x]
    except:
        return x
```

```python
gene_name_dict = region_helpers.gene_id_to_name("/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db")
gene_type_dict = region_helpers.gene_id_to_type("/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db")
```

```python
combined_df['te'] = np.log2(combined_df.riboseq / combined_df.rnaseq)
    
gene_names = []
gene_types = []
for gencode_id in combined_df.index.get_level_values(2):
    try:
        gene_names.append(gene_name_dict[gencode_id])
        gene_types.append(gene_type_dict[gencode_id])
    except KeyError:
        gene_names.append("error")
        gene_types.append("error")
        
        
combined_df["gene_name"] = gene_names
combined_df["gene_type"] = gene_types
combined_df["logged_riboseq"] = combined_df.riboseq.apply(np.log2)
combined_df["logged_rnaseq"] = combined_df.rnaseq.apply(np.log2)
```

```python
with dataviz.Figure(os.path.join(img_dir, "rnaseq_vs_riboseq_heatmap.svg"), figsize=(6,12)) as fig:
    for x, (group, df) in enumerate(combined_df.groupby(level=["cell_type",  'rep'])):
        ax = fig.add_subplot(len(list(combined_df.groupby(level=["cell_type",  'rep']))) / 2 , 2, x +1)
        x = dataviz.loglog_hist2d(df.riboseq, df.rnaseq, ax=ax)
        r2 = scipy.stats.linregress(df.riboseq, df.rnaseq)[2] ** 2
        ax.set_xlabel("Ribo-seq")
        ax.set_ylabel("RNA-seq")
        ax.set_title(" ".join(group) + "\n$R^2${:.2}".format(r2))
```

```python
with dataviz.Figure(os.path.join(img_dir, "rnaseq_vs_riboseq_heatmap.svg"), figsize=(6,12)) as fig:
    gs = gridspec.GridSpec(len(list(combined_df.groupby(level=["cell_type"]))), 2)
    for x, (group, df) in enumerate(combined_df.groupby(level=["cell_type"])):

        rep1, rep2 = df.index.levels[1]
        rep1 = df.ix[group].ix[rep1]
        rep2 = df.ix[group].ix[rep2]
        
        ax = fig.add_subplot(gs[x,0]) 
        rnaseq = pd.concat({"rep1" : rep1.rnaseq, "rep2": rep2.rnaseq}, axis=1).dropna()
        dataviz.loglog_hist2d(rnaseq.rep1, rnaseq.rep2, ax=ax)
        r2 = scipy.stats.linregress(rnaseq.rep1, rnaseq.rep2)[2] ** 2
        spearman = scipy.stats.spearmanr(rnaseq.rep1, rnaseq.rep2)[0]
        ax.set_xlabel("Rep1")
        ax.set_ylabel("Rep2")
        ax.set_title("RNA-seq " + group + "\n$R^2${:.2}\nSpearman{:.2}".format(r2, spearman))
        
        ax = fig.add_subplot(gs[x,1]) 
        
        riboseq = pd.concat({"rep1" : rep1.riboseq, "rep2": rep2.riboseq}, axis=1).dropna()
        dataviz.loglog_hist2d(riboseq.rep1, riboseq.rep2, ax=ax)
        r2 = scipy.stats.linregress(riboseq.rep1, riboseq.rep2)[2] ** 2
        spearman = scipy.stats.spearmanr(riboseq.rep1, riboseq.rep2)[0]

        ax.set_xlabel("Rep1")
        ax.set_ylabel("Rep2")
        ax.set_title("Ribo-seq " + group + "\n$R^2${:.2}\nSpearman {:.2}".format(r2, spearman))
```

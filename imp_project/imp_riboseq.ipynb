```python
half_window_width = 50
def pad(interval):
    interval.start = interval.start - half_window_width 
    interval.stop = interval.stop + half_window_width
    return interval
```

```python
features  = get_genomic_regions.GenomicFeatures( "hg19", 
                                                gffutils.FeatureDB("/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db"),
                                                )
regions = features.get_feature_locations()
padded_start_codons = regions['start_codons'].each(pad).sort().saveas()
padded_stop_codons = regions['stop_codons'].each(pad).sort().saveas()
```

```python
start_distance_df = {}
for experiment in riboseq:
    print "starting ", experiment, 
    start_distance_df[experiment] = riboseq_utils.get_closest(riboseq[experiment], regions['start_codons']).T
    print "done!"
```

```python
#fig = plt.Figure()
with dataviz.Figure(os.path.join(img_dir, "ribosome_occupancy_cdf_start_site.svg"), figsize=(20,10), tight_layout=False) as fig:
    gs = gridspec.GridSpec(2,  len(start_distance_df.items()))
    for count, (name, closest_start_df) in enumerate(start_distance_df.items()):
        closest_start_df_filtered = closest_start_df[(closest_start_df.dist > -30) & (closest_start_df.dist < 30) ]

        ax = fig.add_subplot(gs[0, count])
        x = ax.hist(closest_start_df_filtered.dist, bins=59)
        ax.set_title(name.split("_")[0])
        ax.set_xlabel("Read Start Distance\nfrom known Stop Codon")
        if ax.is_first_col():
            ax.set_ylabel("Count")
        sns.despine(ax=ax)
        
        ax = fig.add_subplot(gs[1, count])
        x = ax.hist(closest_start_df_filtered.length, bins=100, range=(0,100))
        ax.set_xlim(10,40)
        ax.set_xlabel("Fragment Length")
        if ax.is_first_col():
            ax.set_ylabel("Count")
        sns.despine(ax=ax)
```

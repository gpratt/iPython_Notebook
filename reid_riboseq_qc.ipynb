```python
def get_peak_wiggle(tool, data_pos, data_neg, size = 1000, num_peaks=100):
    
    """
    
    returns matrix of (n,25) of wiggle and peak information
    tool - bedtool contains location of peaks
    data - bigwig file handle - contains bw data for the peaks

    """
    
    results = []
    index = []
    for line in tool[:num_peaks]:
        peak_center = line.start
        pad_amt = size / 2
        try:
            if line.strand == "+":
                result = data_pos.get_as_array(line.chrom, peak_center - pad_amt , peak_center + pad_amt)
            if line.strand == "-":
                result = data_neg.get_as_array(line.chrom, peak_center - pad_amt , peak_center + pad_amt)[::-1]
                
            result = [0 if np.isnan(x) else x for x in result]
        
            results.append(result)
            index.append(line.name)
            
        except:
            pass
    return pd.DataFrame(results)
```

```python
def get_length(peak_file):
    if type(peak_file) is float:
        return np.nan
    
    try:
        with open(peak_file) as peak_handle:
            bedtool_size = len(pybedtools.BedTool(peak_handle))
        return bedtool_size
    except ValueError:
        return np.nan
    except IOError:
        return np.nan
    
def get_intervals(peak_file):
    if type(peak_file) is float:
        return np.nan
    try:
        with open(peak_file) as peak_handle:
            bedtool = pybedtools.BedTool(peak_handle)
            intervals = [interval for interval in bedtool]
            return intervals
    except ValueError as e:
        print e
        return np.nan
    except IOError:
        return np.nan

def memory_efficent_intersect(in_tuple, force=True):
    a_bed_file, b_bed_file = in_tuple
    
    if a_bed_file is numpy.nan or b_bed_file is numpy.nan:
        return np.nan
    
    if os.path.getsize(a_bed_file) == 0 or os.path.getsize(b_bed_file) == 0:
        return 0
    
    out_dir = "/nas3/gpratt/projects/public_riboseq/analysis/peak_cashe/intersect/"
    out_name = os.path.join(out_dir, os.path.basename(a_bed_file) + os.path.basename(b_bed_file))
    if os.path.exists(out_name) and not force:
        return out_name
    a_bed_tool = pybedtools.BedTool(a_bed_file)
    b_bed_tool = pybedtools.BedTool(b_bed_file) 
    a_bed_tool.intersect(b_bed_tool, u=True, stream=True, sorted=True).saveas(out_name)
    print out_name

def memory_efficent_union(in_tuple, force=True):
    a_bed_file, b_bed_file = in_tuple

    if a_bed_file is numpy.nan or b_bed_file is numpy.nan:
        return np.nan
    
    if os.path.getsize(a_bed_file) == 0 or os.path.getsize(b_bed_file) == 0:
        return 0
    else:
        
        out_dir = "/nas3/gpratt/projects/public_riboseq/analysis/peak_cashe/union/"
        out_name =  os.path.join(out_dir, os.path.basename(a_bed_file) + os.path.basename(b_bed_file))
        if os.path.exists(out_name) and not force:
            return out_name
        !cat $a_bed_file $b_bed_file | sort -k 1,1 -k 2,2n | cut -f 1-6 | mergeBed > $out_name
    print out_name
    
def make_intersect_file(in_tuple):
    a_bed_file, b_bed_file = in_tuple
    
    if a_bed_file is numpy.nan or b_bed_file is numpy.nan:
        return np.nan
    out_dir = "/nas3/gpratt/projects/public_riboseq/analysis/peak_cashe/intersect/"
    return os.path.join(out_dir, os.path.basename(a_bed_file) + os.path.basename(b_bed_file))

def make_union_file(in_tuple):
    a_bed_file, b_bed_file = in_tuple
    
    if a_bed_file is numpy.nan or b_bed_file is numpy.nan:
        return np.nan
    out_dir = "/nas3/gpratt/projects/public_riboseq/analysis/peak_cashe/union/"
    return os.path.join(out_dir, os.path.basename(a_bed_file) + os.path.basename(b_bed_file))
```

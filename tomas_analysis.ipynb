```python
from collections import Counter
import glob
import os


import pandas as pd
from IPython.core.display import HTML

analysis_dir = "/nas3/gpratt/projects/tomas/"
```

```python
genes = pd.read_excel(os.path.join(analysis_dir, "Wnt Genes.xlsx"), "Sheet1", index_col=1)
```

```python
peak_files = glob.glob("/nas3/gpratt/projects/lin28_missy/v3/*rep*.peaks.bed")
peak_files += glob.glob("/nas3/gpratt/projects/lin28_missy//v3/*.rmDup.sorted.peaks.bed")
peak_files += glob.glob("/nas3/gpratt/projects/public_clip/analysis/v3/*rep*.peaks.bed")
peak_files += glob.glob("/nas3/gpratt/projects/encode/analysis/v7/*rep*.peaks.bed")
peak_files += glob.glob("/nas3/gpratt/projects/public_clip/analysis/v3/*.rmDup.sorted.peaks.bed")
beds = {os.path.basename(peak_file).split(".")[0] : pybedtools.BedTool(peak_file) for peak_file in peak_files}
```

```python
genes_in_beds = {name : Counter([interval.name.split(".")[0] for interval in intervals]) for name, intervals in beds.items()}
```

```python
gene_counts = {}
for name, genes_in_bed in genes_in_beds.items():
    gene_counts[name] = genes.ix[genes_in_bed].dropna()
    gene_counts[name]['counts'] = pd.Series(genes_in_bed, name = "counts")
```

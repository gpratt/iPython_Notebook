```python
from collections import Counter
```

```python
genes_in_beds = {name : Counter([interval.name.split(".")[0] for interval in intervals]) for name, intervals in beds.items()}
```

```python
gene_counts = {}
for name, genes_in_bed in genes_in_beds.items():
    gene_counts[name] = genes.ix[genes_in_bed].dropna()
    gene_counts[name]['counts'] = pd.Series(genes_in_bed, name = "counts")
```

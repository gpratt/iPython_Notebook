```python
import os

from IPython.core.display import HTML
import pandas as pd

from matplotlib_venn import venn3

from gscripts.general import dataviz

fig_dir = "/nas3/gpratt/Dropbox/TAF15/Figures/TAF15_overlap_analysis/"
```

#Code to setup dataframe

```python
master_df = pd.read_excel("/nas3/gpratt/Dropbox/TAF15/Data/striatum_overlap/FETT_clip_plus_halflife.xlsx", "fold_changes", index_col=0)
```

```python
master_df = master_df.drop("Unnamed: 1", axis=1)

master_df['FUS UTR3 clip'][master_df['FUS UTR3 clip'] == "none"] = 0
master_df['TAF UTR3 clip'][master_df['TAF UTR3 clip'] == "none"] = 0
master_df['TDP43 UTR3 clip'][master_df['TDP43 UTR3 clip'] == "none"] = 0
master_df = master_df.astype(float)
master_df['TAF log2 Fold change'] = master_df['TAF Fold change'].apply(log2)

master_df['FUS_3UTR_bound'] = master_df['FUS UTR3 clip'] > 0
master_df['TAF_3UTR_bound'] = master_df['TAF UTR3 clip'] > 0	
master_df['TDP43_3UTR_bound'] = master_df['TDP43 UTR3 clip'] > 0	
```

```python
FUS_up = master_df[master_df['FUS Pvalue or Z-score'] > 1.96]
FUS_down = master_df[master_df['FUS Pvalue or Z-score'] < -1.96]
TAF15_up = master_df[(master_df['TAF log2 Fold change'] > 0) & (master_df['TAF Pvalue or Z-score'] < .05)]
TAF15_down = master_df[(master_df['TAF log2 Fold change'] < 0) & (master_df['TAF Pvalue or Z-score'] < .05)]
TDP_up = master_df[master_df['TDP43 Pvalue or Z-score'] > 1.96]
TDP_down = master_df[master_df['TDP43 Pvalue or Z-score'] < -1.96]
```

#Figure 2 overlaps of up and downregulated TAF, FUS and TDP genes

```python
with dataviz.Figure(os.path.join(fig_dir, "up_and_down_overlaps.svg"), figsize=(10,10)) as fig:
    ax = fig.add_subplot(1,2,1)
    ax.set_title("Upregulated Overlap")
    venn3(subsets=[FUS_up.index, 
                   TAF15_up.index, 
                   TDP_up.index],
          set_labels=["FUS", "TAF15", "TDP43"], ax=ax)
    
    ax = fig.add_subplot(1,2,2)
    ax.set_title("Downregulated Overlap")
    venn3(subsets=[FUS_down.index, 
                   TAF15_down.index, 
                   TDP_down.index],
          set_labels=["FUS", "TAF15", "TDP43"], ax=ax)
```

```python
with dataviz.Figure(os.path.join(fig_dir, "FUS_3UTR_binding.svg"), figsize=(10,10)) as fig:
    ax = fig.add_subplot(1,1,1)
    ax.hist([FUS_up['FUS UTR3 clip'], FUS_down['FUS UTR3 clip'],
      TAF15_up['FUS UTR3 clip'], TAF15_down['FUS UTR3 clip'],
      TDP_up['FUS UTR3 clip'], TDP_down['FUS UTR3 clip']
      ], label=["FUS up", "FUS down",
                "TAF15 up", "TAF15 down",
                "TDP up", "TDP down",],
            bins=10,
            range=(0,10),
            normed=True,
            )
    ax.legend()
    ax.set_xticks(range(10))
    ax.set_title("Histogram  of Number FUS binding sites for different groups")
```

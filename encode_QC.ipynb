```python
from collections import defaultdict, Counter
import glob
import os

from IPython.core.display import HTML, Image
from matplotlib_venn import venn3
import pandas as pd
import pybedtools
import pysam

from gscripts.general import parsers
```

```python
analysis_dir = "/nas3/gpratt/projects/encode/analysis/v7"
combined_df = parsers.clipseq_metrics(analysis_dir, iclip=True)
combined_df = combined_df.rename(columns={"Processed bases" : "Input Bases",
           "Processed reads" : "Input Reads",
           "Number of input reads" : "Reads after Quality Filtering"
           })
```

```python
combined_df["Reads After Triming"] = combined_df['Input Reads'] - combined_df['Too short reads']
filtered_df = combined_df[[ "Input Reads",
                            "Reads After Triming",
                            "Reads after Quality Filtering", 
                            "Uniquely mapped reads number",
                            #"Uniquely mapped reads %",
                            "Usable Reads",
                            "spot",
                            "Num Peaks"
                            ]]


HTML(filtered_df.to_html(formatters={"Input Reads" : parsers.commas,
                                     "Reads After Triming" : parsers.commas,
                                     "Reads after Quality Filtering" : parsers.commas,
                                     "Uniquely mapped reads number" : parsers.commas,
                                     "Usable Reads" : parsers.commas,
                                     "Num Peaks": parsers.commas
                                     } ))
```

```python
HTML(filtered_df.ix[[name for name in combined_df.index if len(name.split(".")[1].split("_")) > 1]].to_html(formatters={"Processed reads" : parsers.commas,
                                     "Reads After Triming" : parsers.commas,
                                     "Number of input reads" : parsers.commas,
                                     "Uniquely mapped reads number" : parsers.commas,
                                     "Usable Reads" : parsers.commas,
                                     "Num Peaks": parsers.commas
                                     }))
```

```python
imp_df = filtered_df.ix[[name for name in combined_df.index if "_IMP" in name]]
imp_df.to_csv("IMP_datasets.csv")
```

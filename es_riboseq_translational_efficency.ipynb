```python
import os

import pandas as pd
import statsmodels.api as sm
import gffutils

from gscripts.general.dataviz import plot_cdf
from gscripts.general.dataviz import Figure

img_dir = "/nas3/gpratt/Dropbox/riboseq_team/figures/es_riboseq_translational_efficency/"
```

```python
riboseq = pd.read_csv("/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_v1/NPC_2_ATCACG_L005_R1.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.rpkm",
            sep="\t",
            index_col=0
            )
rnaseq = pd.read_csv("/nas3/gpratt/projects/es_riboseq/analysis/es_rnaseq_v1/NPC_1_CAGATC_L002_R1.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.rpkm",
            sep="\t",
            index_col=0
            )

riboseq["riboseq"] = riboseq.RPKM
riboseq['rnaseq'] = rnaseq.RPKM
riboseq['te'] = log2(riboseq.riboseq / riboseq.rnaseq)
```

```python
def plot_cdf(cdf_list, **kwargs):
    cdf = sm.distributions.ECDF(cdf_list)
    cdf_linspace = np.linspace(min(cdf_list), max(cdf_list))
    if kwargs['ax'] is not None:
        ax = kwargs['ax']
        del kwargs['ax']
        ax.plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
    else:
        plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
```

```python
with Figure(os.path.join(img_dir, "ribosome_occupancy_cdf.svg"), figsize=(10,10)) as fig:
    ax = fig.add_subplot(1,1,1)
    plot_cdf(riboseq.te[~np.isinf(riboseq.te)].dropna(), ax=ax)
    ax.set_xlabel("log2 Ribosome Occupancy")
    ax.set_ylabel("Cumulative Fraction")
    ax.set_title("NPC Cumulative Fraction of Ribosome Occupancy")
```

```python
human_db = "/nas3/yeolab/Genome/ensembl/gtf/gencode.v17.annotation.gtf.db.old"
db = gffutils.FeatureDB(human_db)
genes = db.features_of_type('gene')
gene_type_dict = {}
gene_name_dict = {}
for gene in genes:
    try:
        gene_type_dict[gene.attributes['gene_id']] = gene.attributes['gene_type']
    except KeyError:
        pass
    
    try:
        gene_name_dict[gene.attributes['gene_id']] = gene.attributes['gene_name']
    except KeyError:
        pass
```

```python
gene_names = []
for gencode_id in riboseq.index:
    try:
        gene_names.append(gene_name_dict[gencode_id])
    except KeyError:
        gene_names.append("error")
riboseq["gene_names"] = gene_names
```

```python
filtered_riboseq = riboseq[~np.isinf(riboseq.te)].dropna()
filtered_riboseq.sort('te', ascending=False)
filtered_riboseq.to_csv("/nas3/gpratt/Dropbox/riboseq_team/NPC_TE.csv")
```

```python
def plot_cdf(cdf_list, **kwargs):
    cdf = sm.distributions.ECDF(cdf_list)
    cdf_linspace = np.linspace(min(cdf_list), max(cdf_list))
    if kwargs['ax'] is not None:
        ax = kwargs['ax']
        del kwargs['ax']
        ax.plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
        ax.set_ylim((0,1))
    else:
        plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
```

```python
#remove ES_1 which is an internal control
def renamer(x):
    if "ES" in x:
        x = x.split("_")
        x[1] = int(x[1]) - 1
        x = "_".join(map(str, x))
    return x
        
```

```python
human_db = "/nas3/yeolab/Genome/ensembl/gtf/gencode.v17.annotation.gtf.db.old"
db = gffutils.FeatureDB(human_db)
genes = db.features_of_type('gene')

gene_type_dict = {}
gene_name_dict = {}
for gene in genes:
    try:
        gene_type_dict[gene.attributes['gene_id']] = gene.attributes['gene_type']
    except KeyError:
        pass
    
    try:
        gene_name_dict[gene.attributes['gene_id']] = gene.attributes['gene_name']
    except KeyError:
        pass
```

```python
riboseq_files = glob.glob("/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_v1/*.rpkm")
rnaseq_files = glob.glob("/nas3/gpratt/projects/es_riboseq/analysis/es_rnaseq_v1/*.rpkm")

riboseq = pd.concat({"_".join(os.path.basename(fn).split("_")[:2]).replace("-", "") : pd.read_csv(fn, sep="\t", index_col=0) for fn in riboseq_files})
rnaseq  = pd.concat({"_".join(os.path.basename(fn).split("_")[:2]).replace("-", "") : pd.read_csv(fn, sep="\t", index_col=0) for fn in rnaseq_files})

riboseq = riboseq.drop("ES_1").rename(renamer)

riboseq["riboseq"] = riboseq.RPKM
riboseq['rnaseq'] = rnaseq.RPKM
riboseq['te'] = log2(riboseq.riboseq / riboseq.rnaseq)
riboseq = riboseq.drop("RPKM", axis=1)

gene_names = []
gene_types = []
for gencode_id in riboseq.index.get_level_values(1):
    try:
        gene_names.append(gene_name_dict[gencode_id])
        gene_types.append(gene_type_dict[gencode_id])
    except KeyError:
        gene_names.append("error")
        gene_types.append("error")
        
        
riboseq["gene_name"] = gene_names
riboseq["gene_type"] = gene_types
riboseq["logged_riboseq"] = riboseq.riboseq.apply(log10)
riboseq["logged_rnaseq"] = riboseq.rnaseq.apply(log10)
riboseq_logged = riboseq[(riboseq.logged_riboseq != np.NINF) & (riboseq.logged_rnaseq != np.NINF) ]
```

#Ribosome Occupancy Descriptive Stats

```python
with Figure(os.path.join(img_dir, "ribosome_occupancy_cdf.svg"), figsize=(10,10)) as fig:
    ax = fig.add_subplot(1,1,1)
    for name in sorted(set(riboseq.index.get_level_values(0))):
        plot_cdf(riboseq.ix[name].te[~np.isinf(riboseq.ix[name].te)].dropna(), ax=ax, linewidth=4, alpha=.75, label=name)
        ax.set_xlabel("log2 Ribosome Occupancy")
        ax.set_ylabel("Cumulative Fraction")
        ax.set_title("Cumulative Fraction of Ribosome Occupancy")
        ax.legend()
```

Currently no major reproducable TE shifts between datasets, but will hopefully change as we get deeper sequencing

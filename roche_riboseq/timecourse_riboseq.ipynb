```python
datasets = glob.glob(os.path.join(riboseq_dir, "*.rpkm"))
datasets = {os.path.basename(dataset).split(".")[0] : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
riboseq_rpkm_df = pd.concat(datasets, join="outer")

datasets = glob.glob(os.path.join(rnaseq_dir, "*.rpkm"))
datasets = {os.path.basename(dataset).split(".")[0] : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
rnaseq_rpkm_df = pd.concat(datasets, join="outer")

master_df = pd.concat([rnaseq_rpkm_df, riboseq_rpkm_df])
master_df = master_df.drop(["cortex_ACAGTG_L004_R1_001", "cortex_GCCAAT_L008_R1_001"], level=0)
master_df = master_df.RPKM.unstack()

new_index = []
for index in master_df.index:
    index = index.split("_")

    sample = index[0]
    lane = int(index[-3][-1])
    exp = "rnaseq" if lane > 5 else "riboseq"
    rep = sample[-1]
    time = sample[:-1]
    if "T" in index:
        time += "_T"
    new_index.append([exp, time, rep])
    
    
master_df.index = pd.MultiIndex.from_tuples(new_index, names=["exp_type", "time", "rep"])

result = []
for exp_name, exp_groups in master_df.groupby(level=['time', "rep"]):
    result.append(exp_groups.ix['riboseq'] / exp_groups.ix['rnaseq'])
te_df = pd.concat(result)

te_df.index = pd.MultiIndex.from_tuples([["te"] + list(cur_index) for cur_index in te_df.index])

master_df = pd.concat([master_df, te_df]).sortlevel()
```

```python
class RiboseqPCA():
    def __init__(self, df):
        """ inits with a multiindex df, combines and classifies"""
        flotilla_df = master_df.copy()
        self.df = df
        df.columns = [gene_name_dict[gene_id] for gene_id in self.df.columns]

        metadata = pd.DataFrame({" ".join(map(str,index)) : index for index in self.df.index}, 
                        #names=["tech_rep", "celltype", "exp"]
                        ).T

        df.index = [" ".join(map(str,index)) for index in self.df.index]
        
        self.classifier = analysis_tools.Comparer('foo', list(self.df.index), self.df.T, metadata, 1)
        self.classifier.fit_classifiers()
        
        
    def plot(self, fig, std):
        self.classifier.score_classifiers(score_cutoff_fun = lambda scores: np.mean(scores) + std*np.std(scores))  

        ax = fig.add_subplot(2,1,1)
        vz = analysis_tools.Visualizer(self.classifier, attributes=(1, 'ExtraTreesClassifier'))
        vz.plot_classifier_scores(ax=ax)
        ax.set_xlim(xmin=0)
        ax.set_ylim(ymin=0)
        ax.set_xlabel('importance')
        trait, classifier_name = vz.attributes
        ax.set_title("%s - %s genes" % ("foo", vz.classifiers[trait][classifier_name].n_good_features), fontsize='small')

        ax = fig.add_subplot(2,1,2)
        vz.do_pca(plotting_args_dict={"ax": ax,
                                      "show_vector_labels": False,
                                      "colors_dict": {index: color_dict[index.split()[0]] for index in self.df.index},
                                      }
                  )
```

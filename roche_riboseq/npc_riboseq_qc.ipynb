#Look at R2 for all technical replicates

```python
datasets = glob.glob(os.path.join(riboseq_dir, "*.rpkm"))
datasets = {os.path.basename(dataset).split(".")[0] : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
riboseq_rpkm_df = pd.concat(datasets, join="outer")

datasets = glob.glob(os.path.join(rnaseq_dir, "*.rpkm"))
datasets = {os.path.basename(dataset).split(".")[0] : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
rnaseq_rpkm_df = pd.concat(datasets, join="outer")

master_df = pd.concat([rnaseq_rpkm_df, riboseq_rpkm_df])
master_df = master_df.RPKM.unstack()

#Each dataset has
#Riboseq / RNAseq (T or R)
#Replicate ID (1,2,3)
#Biorep ID (15,16 (wt)), (17,18, het), (19,20, homo), (21)
new_index = []
for index in master_df.index:
    levels = index.split("_")[0]
    exp_type = "riboseq" if "R" in levels else "rnaseq"
    rep_num = int(levels[0])
    genotype = {'15' : "wt",
                '16' : "wt",
                '17' : "hetrozygous",
                '18' : "hetrozygous",
                '19' : "homozygous",
                '20' : "homozygous",
                '21' : 'homozygous'}[levels[-2:]]
    new_index.append((exp_type, rep_num, genotype, int(levels[-2:])))
    
master_df.index = pd.MultiIndex.from_tuples(new_index, names=["exp_type", "replicate", "genotype", "experiment"])
master_df = master_df.sortlevel()

#calculate translational efficency 
result = []
for exp_name, exp_groups in master_df.groupby(level="experiment"):
    result.append(exp_groups.ix['riboseq'] / exp_groups.ix['rnaseq'])
te_df = pd.concat(result)

te_df.index = pd.MultiIndex.from_tuples([["te"] + list(cur_index) for cur_index in te_df.index])

master_df = pd.concat([master_df, te_df]).sortlevel()
```

```python
with dataviz.Figure(os.path.join(img_dir, "replicate_plotting.svg"), figsize=(10,50)) as fig:
    gs = gridspec.GridSpec(len(list(master_df.groupby(level=("exp_type", "experiment")))), 3)
    for x, (name, group) in enumerate(master_df.groupby(level=("exp_type", "experiment"))):
        for y, (name1, name2) in enumerate(itertools.combinations(group.index, 2)):
            ax = fig.add_subplot(gs[x,y]) 
            series1 = master_df.ix[name1]
            series2 = master_df.ix[name2]

            loglog_hist2d(series1, 
                        series2, 
                        ax=ax)
            sns.despine(ax=ax)
            
            if ax.is_first_col():
                ax.set_ylabel(" ".join(map(str, name)))
            r2 = scipy.stats.linregress(series1, series2)[2]
            ax.set_title("{} vs {}\n{:.2f}".format(name1[1], name2[1], r2))
            
```

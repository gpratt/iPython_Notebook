```python
def get_length(peak_file):
    if type(peak_file) is float:
        return np.nan
    try:
        bedtool = pybedtools.BedTool(peak_file)
        bedtool_size = len(bedtool)
        pybedtools.helpers.close_or_delete(bedtool)
        return bedtool_size
    except ValueError:
        return np.nan

    
size_df = sample_df.applymap(get_length)
```

```python
def get_mean_width(peak_file):
    
    if type(peak_file) is float:
        return np.nan
    try:
        bedtool = pybedtools.BedTool(peak_file)
        bedtool_mean = mean([len(interval) for interval in bedtool])
        pybedtools.helpers.close_or_delete(bedtool)
        return bedtool_mean
    except ValueError as e:
        print e
        return np.nan

    

length_df = sample_df.applymap(get_mean_width)
```

```python
def memory_efficent_intersect(in_tuple):
    a_bed_name, b_bed_name = in_tuple
    if a_bed_name is np.nan or b_bed_name is np.nan:
        return np.nan
    
    if os.path.getsize(a_bed_name) == 0 or os.path.getsize(b_bed_name) == 0:
        return 0
    else:
        intersect = !intersectBed -s -u -a $a_bed_name -b $b_bed_name -sorted | wc -l
        return float(intersect[0])

def memory_efficent_union(in_tuple):
    a_bed_name, b_bed_name = in_tuple
    if a_bed_name is np.nan or b_bed_name is np.nan:
        return np.nan
    
    if os.path.getsize(a_bed_name) == 0 or os.path.getsize(b_bed_name) == 0:
        return 0
    else:
        union = !cat $a_bed_name $b_bed_name | sort -k 1,1 -k 2,2n | cut -f 1-6 | mergeBed -s -nms -scores mean | wc -l
        return float(union[0])
```

```python
pwm_locations = "/nas3/gpratt/projects/cisbp/analysis/pwm_locations/"
def get_closest_general(in_tuple, file_type):
    peak_file, motif_id = in_tuple
    pwm_file = os.path.join(pwm_locations, file_type % (motif_id))
    
    if not os.path.exists(pwm_file):
        return np.nan
            
    if os.path.getsize(pwm_file) == 0 or os.path.getsize(peak_file) == 0:
        return np.nan
    
    peak_tool = pybedtools.BedTool(peak_file)
    pwm_tool = pybedtools.BedTool(pwm_file) 
    return peak_tool.closest(pwm_tool, s=True, D="a", t="first").saveas()

def get_closest(in_tuple):
    get_closest_general(in_tuple, "%s.txt/fimo_reformated.txt")
    
def get_closest_shuffled(in_tuple):
    get_closest_general(in_tuple, "%s.txt/fimo_reformated.shuffled.txt")
```

```python
closest_df = pwm_overlap_df.applymap(get_closest)
closest_shuffled_df = pwm_overlap_df.applymap(get_closest_shuffled)
```

```python
def get_percent_overlapped(tool, tool_type):
    
    """
    gets number of features that overlap a motif sorted by signifiance (the score column)
        returns x and y for plotting
    """
    
    total_overlapped = 0
    percent_overlapped = []
    sorted_tool = sorted(tool, key=lambda x: float(x[4]))
    for total_counted, interval in enumerate(sorted_tool):
        if int(interval[-1]) == 0: #there is an overlap
            total_overlapped += 1.0
        percent_overlapped.append((total_overlapped / (total_counted + 1)) * 100)
    return range(len(sorted_tool)), percent_overlapped
```

```python
import os
import glob
from collections import Counter
from itertools import groupby

from IPython.core.display import Image
from bx.bbi.bigwig_file import BigWigFile 
import pandas as pd
import pysam
from sklearn import svm
from sklearn import cross_validation
from sklearn import metrics

from clipper.src import get_genomic_regions
from gscripts.general import dataviz
from gscripts.riboseq.uorf_detector import UORF_detector
img_dir = "/nas3/gpratt/Dropbox/riboseq_team/figures/es_riboseq_start_site_predictor/"
analysis_dir = "/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_v1"
```

#Code

```python
def true_pad(interval):
    if interval.strand == "+":
        interval.stop = interval.stop
        interval.start = interval.start - 3
    else:
        interval.start = interval.start - 3 
        interval.stop = interval.stop 
    interval.start = interval.start - 15
    interval.stop = interval.stop + 15
    if len(interval) != 34:
        return None
    return interval

def random_pad(interval, downstream=70):
    if interval.strand == "-":
        downstream = downstream * -1
        
    interval.start = interval.start + downstream   
    interval.stop = interval.stop + downstream
    interval.start = interval.start - 33
    if len(interval) != 34:
        return None
    return interval
```

```python
def get_profile(interval):
    if interval.strand == "+":
        profile = pos_bw.get_as_array(interval.chrom, interval.start, interval.stop)
    else:
        profile = neg_bw.get_as_array(interval.chrom, interval.start, interval.stop)[::-1] * -1
    profile[isnan(profile)] = 0
    return profile
```

#Gather data, bigwig files, read densities and predicted start sites

```python
genomic_features = get_genomic_regions.GenomicFeatures("hg19", None)
regions = genomic_features.get_feature_locations()
```

```python
neg_bw, pos_bw = [BigWigFile(open(fn)) for fn in glob.glob(os.path.join(analysis_dir, "NPC_2_ATCACG_L005_R1*neg.t.bw")) +
glob.glob(os.path.join(analysis_dir, "NPC_2_ATCACG_L005_R1*pos.bw"))]
```

```python
riboseq_expression = {os.path.basename(name.split(".")[0]) : 
           pd.read_csv(name, sep="\t", index_col=0) for name in glob.glob(os.path.join(analysis_dir, "*.rpkm"))}
```

```python
expressed_genes = riboseq_expression['NPC_2_ATCACG_L005_R1'][riboseq_expression['NPC_2_ATCACG_L005_R1'].RPKM > 10].index
expressed_starts = regions['start_codons'].filter(lambda x: x.name in expressed_genes).saveas()
```

```python
padded_start_codons = expressed_starts.each(true_pad).saveas()
random_padded_start_codons = expressed_starts.each(random_pad, downstream=70).saveas()
```

```python
normal_expressed_array = array([get_profile(interval) for interval in padded_start_codons])
random_expressed_array = array([get_profile(interval) for interval in random_padded_start_codons]) 
```

#Plot Descriptives

```python
with dataviz.Figure(os.path.join(img_dir, "read_distribution.svg"), figsize=(6,6)) as fig:
    ax = fig.add_subplot(1,1,1)
    ax.plot(normal_expressed_array.sum(axis=0), label="Real", linewidth=3)
    ax.plot(random_expressed_array.sum(axis=0), label="Random", linewidth=3)
    ax.set_ylabel("Cumulative Number of Reads")
    ax.set_xlabel("Distribution Around Start\nCodon (codon in center) (bp)")
    ax.set_title("Meta Gene Analysis\nof Real and Random Start Sites")
    ax.legend(frameon=False)
```

We see a clear spike near the start site (center of figure, although there is an odd, and unexpected distribution for the random data.  Random being generated as 70bp downstream of start codons.

#Get Percision and Recall from training data

```python
X = concatenate([normal_expressed_array, random_expressed_array])
Y = array(["True"] * len(normal_expressed_array) + ["False"] * len(random_expressed_array))
```

```python
true = []
test = []
C = 1.0
for train_index, test_index in cross_validation.ShuffleSplit(len(X), test_size=.1, indices=False):
    svc = svm.SVC(kernel="linear", C=C).fit(X[train_index],Y[train_index])
    true.append(Y[test_index])
    test.append(svc.predict(X[test_index]))
    
```

```python
percision = []
recall = []
for true_y, test_y in zip(true, test):
    percision.append(metrics.precision_score(true_y, test_y, pos_label="True"))
    recall.append(metrics.recall_score(true_y, test_y, pos_label="True"))
```

```python
with Figure(os.path.join(img_dir, "percision_recall.svg"), figsize=(6,6)) as fig:
    ax = fig.add_subplot(1,1,1)
    ax.boxplot([percision, recall])
    ax.set_title("Percision and Recall\nfor 10 fold cross validation")
    ax.set_xticklabels(["Percision", "Recall"])
    ax.set_ylim((0,1))
```

Very few false positives, but I have a large number of false negatives, leading to a fairly high miss-classification rate...

```python
def get_densities(lst, data_pos, data_neg):
    result  = []
    for interval in lst:
        if len(interval) == 0:
            return []
        if interval.strand == "+":
            interval_scores = data_pos.get_as_array(interval.chrom, interval.start, interval.end)
        if interval.strand == "-":
            interval_scores = data_neg.get_as_array(interval.chrom, interval.start, interval.end)[::-1] * -1
        interval_scores[isnan(interval_scores)] = 0 
        result.append(interval_scores)

    return result 
```

```python
def generate_region_dict(bedtool):
    region_dict = defaultdict(list)
    
    for interval in bedtool:
        region_dict[interval.name].append(interval)
    
    for gene in region_dict.keys():
        if region_dict[gene][0].strand == "-":
            region_dict[gene].reverse()
            
    return region_dict
```

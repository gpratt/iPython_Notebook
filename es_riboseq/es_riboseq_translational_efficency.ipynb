```python
def plot_cdf(cdf_list, **kwargs):
    cdf = sm.distributions.ECDF(cdf_list)
    cdf_linspace = np.linspace(min(cdf_list), max(cdf_list))
    if kwargs['ax'] is not None:
        ax = kwargs['ax']
        del kwargs['ax']
        ax.plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
        ax.set_ylim((0,1))
    else:
        plot(cdf_linspace, cdf(cdf_linspace), **kwargs)
```

```python
#remove ES_1 which is an internal control
def renamer(x):
    if "ES" in x:
        x = x.split("_")
        x[1] = int(x[1]) - 1
        x = "_".join(map(str, x))
    return x
        
```

#Setup data frame

```python
human_db = "/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db"
db = gffutils.FeatureDB(human_db)
genes = db.features_of_type('gene')

gene_type_dict = {}
gene_name_dict = {}
for gene in genes:
    try:
        gene_type_dict[gene.attributes['gene_id'][0]] = gene.attributes['gene_type'][0]
    except KeyError:
        pass
    
    try:
        gene_name_dict[gene.attributes['gene_id'][0]] = gene.attributes['gene_name'][0]
    except KeyError:
        pass
```

```python
es_color, npc_color, sox17_color, mesoderm_color, c1 = map(matplotlib.colors.rgb2hex, seaborn.color_palette('Set1', 5))
```

```python
descriptors = defaultdict(dict)
for item in set(riboseq.index.get_level_values(0)):
    if "ES" in item:
        descriptors[item] = {"cell_type" : "ES", "cell_color" : es_color}
    if "NPC" in item:
        descriptors[item] = {"cell_type" : "NPC", "cell_color" : npc_color}
        
    if "SOX17" in item:
        descriptors[item] = {"cell_type" : "SOX17", "cell_color" : sox17_color}
        
    if "TGFP" in item:
        descriptors[item] = {"cell_type" : "TGFP", "cell_color" : mesoderm_color}
        
descriptors = pd.DataFrame(descriptors).T
```

```python
riboseq_files = glob.glob("/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_v1/*.rpkm")
rnaseq_files = glob.glob("/nas3/gpratt/projects/es_riboseq/analysis/es_rnaseq_v1/*.rpkm")

riboseq_dict = {"_".join(os.path.basename(fn).split("_")[:2]).replace("-", "") : pd.read_csv(fn, sep="\t", index_col=0) for fn in riboseq_files}
rnaseq  = pd.concat({"_".join(os.path.basename(fn).split("_")[:2]).replace("-", "") : pd.read_csv(fn, sep="\t", index_col=0) for fn in rnaseq_files})

riboseq_r2_files = glob.glob("/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_second_round_v1//*.rpkm")
riboseq_r2_dict  = {"_".join(os.path.basename(fn).split("_")[:2]) + "_r2": pd.read_csv(fn, sep="\t", index_col=0) for fn in riboseq_r2_files}

riboseq_dict.update(riboseq_r2_dict)
riboseq = pd.concat(riboseq_dict)

riboseq = riboseq.drop("ES_1").rename(renamer)
riboseq = riboseq.rename({"NPC1_Ribo" : "NPC_1"})

riboseq["riboseq"] = riboseq.RPKM
riboseq['rnaseq'] = rnaseq.RPKM
riboseq['te'] = log2(riboseq.riboseq / riboseq.rnaseq)
riboseq = riboseq.drop(["RPKM", "flag"], axis=1)

gene_names = []
gene_types = []
for gencode_id in riboseq.index.get_level_values(1):
    try:
        gene_names.append(gene_name_dict[gencode_id])
        gene_types.append(gene_type_dict[gencode_id])
    except KeyError:
        gene_names.append("error")
        gene_types.append("error")
        
        
riboseq["gene_name"] = gene_names
riboseq["gene_type"] = gene_types
riboseq["logged_riboseq"] = riboseq.riboseq.apply(log2)
riboseq["logged_rnaseq"] = riboseq.rnaseq.apply(log2)
riboseq_logged = riboseq[(riboseq.logged_riboseq != np.NINF) & (riboseq.logged_rnaseq != np.NINF) ]
```

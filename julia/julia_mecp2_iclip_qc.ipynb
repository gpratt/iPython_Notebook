```python
iclip_metrics = parsers.clipseq_metrics("/nas3/gpratt/projects/julia/analysis/mecp2_iclip_v2/", iclip=True)
iclip_metrics = iclip_metrics.rename(columns={"Processed bases" : "Input Bases",
           "Processed reads" : "Input Reads",
           "Number of input reads" : "Reads after Quality Filtering"
           })

iclip_metrics["percent_repetative"] = 1 - (iclip_metrics['Reads after Quality Filtering'] / iclip_metrics['Input Reads'].astype(float))
iclip_metrics["Reads After Triming"] = iclip_metrics['Input Reads'] - iclip_metrics['Too short reads']
iclip_metrics["%usable / input"] = (iclip_metrics['Usable Reads'] / iclip_metrics['Uniquely mapped reads number']) * 100
iclip_metrics["%usable / mapped"] = (iclip_metrics['Usable Reads'] / iclip_metrics['Input Reads']) * 100
iclip_metrics = iclip_metrics.drop(["Finished on", 
                  "Mapping speed, Million of reads per hour", 
                  "Trimmed bases",
                  "Started job on",
                  "Started mapping on"], axis=1)
```

```python
filtered_df = iclip_metrics[[ "Input Reads",
                            "Reads After Triming",
                            "Reads after Quality Filtering", 
                            "Uniquely mapped reads number",
                            #"Uniquely mapped reads %",
                            "Usable Reads",
                            #"spot",
                            "Num Peaks",
                            "%usable / input",
                            "%usable / mapped"
                            
                            ]]

HTML(filtered_df.to_html(formatters={"Input Reads" : parsers.commas,
                                     "Reads After Triming" : parsers.commas,
                                     "Reads after Quality Filtering" : parsers.commas,
                                     "Uniquely mapped reads number" : parsers.commas,
                                     "Usable Reads" : parsers.commas,
                                     "Num Peaks": parsers.commas,
                                     } ))
```

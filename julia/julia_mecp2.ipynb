```python
from itertools import groupby, combinations
import glob
import os
from collections import defaultdict
import gffutils
import scipy.stats
from IPython.core.display import HTML
import pandas as pd

from gscripts.general import parsers
from gscripts.GO import hg19GO
```

```python
human_db = "/nas3/gpratt/gencode/gencode.v17.annotation.gtf.db"
db = gffutils.FeatureDB(human_db)
genes = db.features_of_type('gene')
gene_type_dict = {}
gene_name_dict = {}
gene_len_dict  = {} 
for gene in genes:
    gene_id = gene.attributes['gene_id'][0] if type(gene.attributes['gene_id']) == list else gene.attributes['gene_id']
    try:      
        gene_type_dict[gene_id] = gene.attributes['gene_type'][0]
    except KeyError:
        pass
    
    try:
        gene_name_dict[gene_id] = gene.attributes['gene_name'][0]
    except KeyError:
        pass
    
    try:
        gene_len_dict[gene_id] = len(gene)
    except KeyError:
        pass
gene_id_dict = {value[0] : name for name, value in gene_name_dict.items()}

gene_type_dict = defaultdict(str, gene_type_dict.items())
gene_name_dict = defaultdict(str, gene_name_dict.items())
```

```python
datasets = glob.glob("/nas3/gpratt/projects/julia/analysis/mecp2_ripseq_v1/*.rpkm")


datasets = {os.path.basename(dataset) : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
ripseq_df = pd.concat(datasets, join="outer")
ripseq_df = ripseq_df.drop("flag",1)
ripseq_df['gene_id'] = ripseq_df.index.get_level_values(1)
ripseq_df['gene_name'] = ripseq_df.gene_id.apply(lambda x: gene_name_dict[x])
ripseq_df['gene_type'] = ripseq_df.gene_id.apply(lambda x: gene_type_dict[x])
ripseq_df = ripseq_df.drop("gene_id",1)

renamer = lambda item: "_".join(item.split("_")[1:5])
ripseq_df = ripseq_df.rename({"B4P22_MeCP2_input_3_GCCAAT_L006" : "B4P22_MeCP2_RIP_Input_9_GCCAAT_L006",
                 "B4P22_MeCP2_IP_3_CTTGTA_L006" : "B4P22_MeCP2_RIP_IP_9_CTTGTA_L006",
                 "B4P24_MeCP2_input_8_CGATGT_L006" : "B4P24_MeCP2_RIP_Input_8_CGATGT_L006",
                 "B4P24_MeCP2_IP_8_TGACCA_L006" : "B4P24_MeCP2_RIP_IP_8_TGACCA_L006"})

ripseq_df.index = ripseq_df.index.set_levels([[renamer(name) for name in names] if i==0 else names for i, names in enumerate(ripseq_df.index.levels)])
```

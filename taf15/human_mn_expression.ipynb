```python
splicing_matrix = human_mn_splicing.miso_posterior_mean.unstack()
```

```python
metadata = {}
for item in splicing_matrix.index:
    experiment = item.split("_")[:-2]
    if len(experiment) == 0:
        experiment = item.split("_")[:-1]
    experiment = "_".join(experiment)
    replicate = item.split("_")[-1]
    
    outlier = False
    if item == "FUS_TAF15_shRNA_2":
        outlier = True
    metadata[item]= {"phenotype": experiment, "rep": replicate, 
                     #"outlier": outlier
                     }
metadata = pd.DataFrame(metadata).T
metadata.index.name = "exp"
```

```python
expression = pd.concat({os.path.basename(item).split(".")[0]: pd.read_table(item, index_col=0) for item in glob.glob("/nas3/gpratt/projects/fet_family/analysis/human_mn_v1/*.rpkm")})
```

```python
result = []
for key, values in metadata.iterrows():
    result.append(list(values.values) + [key])
    
expression_index = pd.MultiIndex.from_tuples(result, names=["knockdown", "rep", "exp"])
expression_matrix = expression.RPKM.unstack()
expression_matrix[expression_matrix == 0] = np.nan
expression_matrix = expression_matrix.dropna(axis=1)
expression_matrix.columns = [gene_id.split(".")[0] for gene_id in expression_matrix.columns]
```

```python
metadata.to_csv("/nas3/gpratt/Dropbox/TAF15/Data/MN_data/metadata.csv")
```

```python
study = flotilla.Study(# The metadata describing phenotype and pooled samples
                       metadata,
                       # A version for this data
                       version='0.1.0',
                       expression_data = expression_matrix,
                       # Dataframe of the splicing data
                       splicing_data=splicing_matrix,
                       expression_thresh = 1,
                       species="hg19"
)

```

```python
with dataviz.Figure(os.path.join(img_dir, "replicate_plotting.svg"), figsize=(7.5,5.0)) as fig:
    max_combinations = np.max(expression_matrix.groupby(level=("knockdown")).count().values)
    gs = gridspec.GridSpec(2, 3 
                           #len(list(expression_matrix.groupby(level=("knockdown")))), 
                           #len(list(itertools.combinations(range(max_combinations), 2)))
                           )
    count = 0
    for x, (name, group) in enumerate(expression_matrix.groupby(level=("knockdown"))):
        for y, (name1, name2) in enumerate(itertools.combinations(group.index, 2)):
            
            x = count % 2
            y = count / 2

            ax = fig.add_subplot(gs[x,y]) 
            series1 = expression_matrix.ix[name1]
            series2 = expression_matrix.ix[name2]

            dataviz.loglog_hist2d(series1, series2, ax=ax)
            sns.despine(ax=ax)
            name1 = " ".join(name1[-1].split("_"))
            name2 = " ".join(name2[-1].split("_"))
            
            ax.set_xlabel(name1, fontsize=10)
            ax.set_ylabel(name2, fontsize=10)            
            r2 = scipy.stats.linregress(series1, series2)[2]
            ax.text(2, 3, "$R^2$={:.2f}".format(r2))
            ax.set_title("{} vs\n{}".format(name1, name2), fontsize=6)
            [tick.set_fontsize(10) for tick in ax.get_xticklabels()]
            [tick.set_fontsize(10) for tick in ax.get_yticklabels()]
            count += 1
```

```python
change_direction = Counter()
incorrect = []
for index, sig_item in significant_splicing_events.groupby(level=['rbp', 'splice_type', 'event']):
    if len(sig_item) >= 4:
        if all([event > 0 for event in sig_item['diff']]):
            change_direction['correct_incl'] += 1
        elif  all([event < 0 for event in sig_item['diff']]):
            change_direction['correct_excl'] += 1
        else:
            change_direction['incorrect'] += 1
            incorrect.append(sig_item)
```

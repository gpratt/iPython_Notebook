The first two columns are my RBPs and rest are the mitochondiral Genes.  There is low variance which isn't encouraging, but I'll try to see if I take a "knocked down group" and an "over expressed" group and see how other genes change

```python
knockdown = mn_mito_df[(mn_mito_df['ENSG00000172660'] < 1) & (mn_mito_df['ENSG00000089280'] < 3) ]
over_expressed = mn_mito_df[(mn_mito_df['ENSG00000172660'] > 5) & (mn_mito_df['ENSG00000089280'] > 4) ]
```

#Pseudo-knockdown Analysis

```python
with dataviz.Figure(os.path.join(img_dir, "taf15_fus_correlation.svg"), figsize=(15, 5), tight_layout=False) as fig:
    ax = fig.add_subplot(1,2,1)
    sns.violinplot(knockdown, ax=ax)
    ax.set_title("Knockdown")
    ax.set_ylim(0,20)
    
    ax = fig.add_subplot(1,2,2)
    sns.violinplot(over_expressed)
    ax.set_title("Over Expressed")
    ax.set_ylim(0,20)
```

```python
pd.concat({"knockdown": knockdown.mean(), "over_expressed": over_expressed.mean()}, axis=1)
```

If we were validating out the MN results we would hope that Knockdowns have a higher expression than the over expressed cells.  This is not the case. As visible via both the violin plots and looking at the means

#Trying regression on all the genes, first plot, gene by gene the correlatiom

```python
num_cols = len(mn_mito_df.columns[:2])
num_rows = len(mn_mito_df.columns[2:])
with dataviz.Figure(os.path.join(img_dir, "taf15_fus_correlation.svg"), figsize=(num_cols * 2.5, num_rows * 2.5)) as fig:
    gs = gridspec.GridSpec(num_rows, num_cols)
    for x, row in enumerate(mn_mito_df.columns[:2]):
        for y, col in enumerate(mn_mito_df.columns[2:]):    
            ax = fig.add_subplot(gs[y,x]) 
            ax.scatter(mn_mito_df[row], mn_mito_df[col], alpha=.7)
            if ax.is_first_row():
                ax.set_title(row)
            ax.set_ylabel(col)
            #ax.set_title(row + " " + col, fontsize=8)
            r2 = scipy.stats.linregress(mn_mito_df[row], mn_mito_df[col])[2]
            ax.text(2, 3, "$R^2$={:.2f}".format(r2))
```

```python
result = {}
for gene in mn_mito_df[mn_mito_df.columns[2:]]:
    regr = linear_model.LinearRegression()
    result[gene] = regr.fit(mn_mito_df[mn_mito_df.columns[:2]].values, mn_mito_df[gene])
```

```python
for key, value in result.items():
    print key, value.coef_
```

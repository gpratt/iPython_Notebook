#Get only MN single cells

```python
significant_df = {}
for celltype in set(study.metadata.data.celltype):
    print len(study.sample_subset_to_sample_ids(celltype))
    expression = study.expression.data.ix[study.sample_subset_to_sample_ids(celltype)]
    significant_df[celltype] = expression.dropna()
significant_df = pd.concat(significant_df, names=["celltype", "id"])
```

#Get only significantly chagning mito genes and my RBPs of interest

```python
mito_df = significant_df[[u'ENSG00000172660', 
                            'ENSG00000089280', 
                            'ENSG00000198695',
                             'ENSG00000198727',
                             'ENSG00000198763',
                             'ENSG00000198786',
                             'ENSG00000198886',
                             'ENSG00000198888',
                             'ENSG00000198899',
                             'ENSG00000212907',
                             'ENSG00000228253']]

mito_df = pd.DataFrame(mito_df.stack())
mito_df.index.names = ["celltype", 'id', 'gene_id']
mito_df.columns = ['rpkm']
mito_df['celltype'] = mito_df.index.get_level_values("celltype")
mito_df['gene_id'] = mito_df.index.get_level_values("gene_id")
```

#Whats the general distribution of these genes?

```python
with dataviz.Figure(os.path.join(img_dir, "taf15_fus_correlation.svg"), figsize=(10, 10)) as fig:
    for x, (name, df) in enumerate(mito_df.groupby(level="celltype")):

        ax = fig.add_subplot(2, 2, x + 1)
        sns.violinplot(df.rpkm, groupby=df.gene_id, ax=ax)
        ax.set_title(name)
        ax.set_xticklabels(map(study.expression.feature_renamer, [gene_id.get_text() for gene_id in list(ax.get_xticklabels())]))
        [tick.set_rotation(45) for tick in ax.get_xticklabels()]
```

The first two columns are my RBPs and rest are the mitochondiral Genes.  There is low variance which isn't encouraging, but I'll try to see if I take a "knocked down group" and an "over expressed" group and see how other genes change

```python
knockdown = []
over_expressed = []
for x, df in mito_df.groupby(level=['celltype', "id"]):
    if  df.xs("ENSG00000172660", level="gene_id").rpkm[0] < 1 and df.xs("ENSG00000089280", level="gene_id").rpkm[0] < 4:
        knockdown.append(df)
    elif df.xs("ENSG00000172660", level="gene_id").rpkm[0] > 5 and df.xs("ENSG00000089280", level="gene_id").rpkm[0] > 4:
        over_expressed.append(df)
        
knockdown = pd.concat(knockdown)
over_expressed = pd.concat(over_expressed)
```

#Pseudo-knockdown Analysis

```python
num_rows = 4
num_cols = 2
with dataviz.Figure(os.path.join(img_dir, "taf15_fus_correlation.svg"), figsize=(3.5 * num_cols, 3.0 * num_rows)) as fig:
    gs = gridspec.GridSpec(num_rows, num_cols)
    for x, ((name, knockdown_df), (name, over_expressed_df)) in enumerate(izip(knockdown.groupby(level="celltype"), over_expressed.groupby("celltype"))):
        ax = fig.add_subplot(gs[x, 0])
        sns.violinplot(knockdown_df.rpkm, 
                       groupby=knockdown_df.gene_id, 
                       inner="points",
                       ax=ax)
        ax.set_title("Knockdown " + name)
        ax.set_ylim(0,20)
        ax.set_xticklabels(map(study.expression.feature_renamer, [gene_id.get_text() for gene_id in list(ax.get_xticklabels())]))
        [tick.set_rotation(45) for tick in ax.get_xticklabels()]

        ax = fig.add_subplot(gs[x, 1])
        sns.violinplot(over_expressed_df.rpkm, 
                       groupby=over_expressed_df.gene_id, 
                       ax=ax)
        ax.set_title("Over Expressed")
        ax.set_ylim(0,20)
        ax.set_xticklabels(map(study.expression.feature_renamer, [gene_id.get_text() for gene_id in list(ax.get_xticklabels())]))
        [tick.set_rotation(45) for tick in ax.get_xticklabels()]
```

#Do a T-test between the OE condition and the knockdown condition to see if things are significantly different

```python
significant_changing = {}
for (name_1, knockdown_df), (name_2, over_expressed_df) in izip(knockdown.groupby(level=["celltype", "gene_id"]), over_expressed.groupby(level=["celltype", "gene_id"])):
    t_test = stats.ttest_ind(knockdown_df.rpkm, over_expressed_df.rpkm)
    significant_changing[name_1] =  {"t": t_test[0], "p": t_test[1]}
significant_changing = pd.DataFrame(significant_changing).T
significant_changing.index.names =["celltype", "gene_id"]
significant_changing['gene_name'] = map(study.expression.feature_renamer, significant_changing.index.get_level_values(level="gene_id"))
```

#Correlation Between Technical Replicates

```python
rpkms = {os.path.basename(name).split(".")[0]: pd.read_table(name, index_col=0).RPKM for name in glob.glob("/nas3/gpratt/projects/fet_family/analysis/fet_family_v1/*.rpkm")}
rpkms = pd.concat(rpkms).unstack()

#Straitum
rpkms = rpkms.rename({"KK1": "HTT_1_1_control",
              "KK2": "HTT_2_1_control",
              "KK3": "HTT_3_1_control",
              "KK4": "TIA1_1_1_ASO",
              "KK5": "TIA1_2_1_ASO",
              "KK6": "TIA1_3_1_ASO",
              "KK7": "TAF15_1_1_ASO",
              "KK8": "TAF15_2_1_ASO",
              "KK9": "TAF15_3_1_ASO",
})

rpkms = rpkms.drop(["20090209_MP_none1_mouse_brain_CLIPseq_TDP43_lower_band",
                     "20091230_MP10_1_TLS_clip_mouse_brain_3A",
                     "20091230_MP11_1_TLS_clip_mouse_brain_3B"])

new_index = []
for index in rpkms.index:
    if "TDP43" in index:
        gene = "TDP43"
    elif "TLS" in index:
        gene = "TLS"
    elif "TIA1" in index:
        gene = "TIA1"
    elif "TAF15" in index:
        gene = "TAF15"
    elif "control" in index:
        gene = "HTT Control"
    elif "embryonic_brain_WT" in index:
        gene = "Embryonic Brain Control"
    elif "Saline" in index:
        gene = "Saline Control"
        
    if "Liver" in index:
        cell_type = "liver"
    elif "embryonic_brain" in index:
        cell_type = "embryonic_brain"
    else:
        cell_type = "striatum" 
        
    reps = index.split("_")
    biological_replicate = reps[1]
    technical_replicate = reps[2]
    new_index.append((gene, cell_type, biological_replicate, technical_replicate, index))
    
rpkms.index = pd.MultiIndex.from_tuples(new_index, names=["rbp", 
                                                          "cell_type", 
                                                          "biological_replicate",
                                                          "technical_replicate",
                                                          "sample"])
rpkms = rpkms.sortlevel()
```

```python
with dataviz.Figure(os.path.join(img_dir, "replicate_plotting.svg"), figsize=(10,50)) as fig:
    max_combinations = np.max(rpkms.groupby(level=("rbp", "cell_type")).count().values)
    gs = gridspec.GridSpec(len(list(rpkms.groupby(level=("rbp", "cell_type", "biological_replicate")))), 
                           len(list(itertools.combinations(range(max_combinations), 2)))
                           )
    for x, (name, group) in enumerate(rpkms.groupby(level=("rbp", "cell_type", "biological_replicate"))):
        for y, (name1, name2) in enumerate(itertools.combinations(group.index, 2)):
            ax = fig.add_subplot(gs[x,y]) 
            series1 = rpkms.ix[name1]
            series2 = rpkms.ix[name2]

            dataviz.loglog_hist2d(series1, 
                        series2, 
                        ax=ax)
            sns.despine(ax=ax)
            
            if ax.is_first_col():
                ax.set_ylabel(" ".join(map(str, name)))

            #r2 = scipy.stats.linregress(series1, series2)[2]
            ax.set_title("{} vs {}\n{:.2f}".format(name1[-1], name2[-1], 
                                                   #r2
                                                   )
                         )
        print "done"
```

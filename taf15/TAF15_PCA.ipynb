```python
import gffutils
import matplotlib
import pandas as pd
import seaborn

from gscripts.general.analysis_tools import Comparer, Visualizer
from gscripts.general.dataviz import PCA_viz

from gscripts.GO import mm9GO as GO
```

```python
human_db = "/nas3/gpratt/gencode/gencode.vM1.annotation.gtf.db"
db = gffutils.FeatureDB(human_db)
genes = db.features_of_type('gene')

gene_type_dict = {}
gene_name_dict = {}
for gene in genes:
    try:
        gene_type_dict[gene.attributes['gene_id'][0]] = gene.attributes['gene_type'][0]
    except KeyError:
        pass
    
    try:
        gene_name_dict[gene.attributes['gene_id'][0].split(".")[0]] = gene.attributes['gene_name'][0]
    except KeyError:
        pass
```

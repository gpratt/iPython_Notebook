```python
import os
import glob
from collections import Counter
from itertools import groupby

from IPython.core.display import Image
from bx.bbi.bigwig_file import BigWigFile 
import pandas as pd
import pysam

from clipper.src import CLIP_analysis
from gscripts.general.dataviz import Figure

img_dir = "/nas3/gpratt/Dropbox/riboseq_team/figures/es_riboseq_rpf/"
analysis_dir = "/nas3/gpratt/projects/es_riboseq/analysis/es_riboseq_v1"
```

#Code

```python
def get_five_prime_end(interval):
    interval.score = str(sum(map(int, interval[10].split(","))))
    if interval.strand == "+":
        interval.stop = interval.start
    else:
        interval.start = interval.stop
    return interval
```

```python
def pad(interval):
    interval.start = interval.start - 50 
    interval.stop = interval.stop + 50
    return interval
```

#Gather Data

```python
regions = CLIP_analysis.get_feature_locations("/nas3/gpratt/clipper/clipper/data/regions/", "hg19", None)
padded_start_codons = regions['start_codons'].each(pad).saveas()
padded_stop_codons = regions['stop_codons'].each(pad).saveas()
```

```python
riboseq = {os.path.basename(name.split(".")[0]) : 
           pybedtools.BedTool(name) for name in glob.glob(os.path.join(analysis_dir, "*.rg.bam"))}
```

```python
start_codon_overlapping_reads = riboseq["NPC_2_ATCACG_L005_R1"].intersect(padded_start_codons, 
                                                                          u=True, 
                                                                          s=True, 
                                                                          sorted=True).saveas().saveas("start_overlapping_reads.bam")
start_codon_reads_bed = start_codon_overlapping_reads.bam_to_bed(bed12=True).saveas()
start_codon_true_starts = start_codon_reads_bed.each(get_five_prime_end).saveas()
closest_start_codons = start_codon_true_starts.closest(regions['start_codons'], D="b", s=True).saveas()
closest_start_df = pd.DataFrame({x.name : {"dist" : int(x[-1]), "length" : int(x.score)} for x in closest_start_codons}).T
closest_start_df_filterd = closest_start_df[(closest_start_df.dist > -30) & (closest_start_df.dist < 0) ]
```

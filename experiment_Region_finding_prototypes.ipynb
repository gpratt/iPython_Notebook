```python
#this annotates potental NMD Genes 

#gets all transcripts and their exons and lines them up in a list
transcript_dict = defaultdict(list)
exons = gff.filter(lambda x: x[2] == "exon")
for exon in exons:
    transcript_dict[exon.attrs['Parent']].append(exon)

#looks for the location of each stop codon in the transcript and exon of its parent, annotates it with NMD important information
stop_codons = gff.filter(lambda x: x[2] == "stop_codon") 
ejc_annotations = {}
for stop_codon in stop_codons:
    transcript_id = stop_codon.attrs['Parent']
    exons = transcript_dict[transcript_id]
    for exon_num, exon in enumerate(exons):
        
        #then the stop codon is in the exon...
        if stop_codon.start > exon.start and stop_codon.start < exon.stop:
            if stop_codon.strand == "+":
                distance_from_three_prime_end = stop_codon.start - exon.stop
                num_ejcs_remaining = len(exons) - (exon_num + 1)
            elif stop_codon.strand == "-":
                distance_from_three_prime_end = exon.start - stop_codon.start
                num_ejcs_remaining = exon_num
    ejc_annotations[transcript_id] = {}
    ejc_annotations[transcript_id]['distance_from_three_prime_end'] = distance_from_three_prime_end
    ejc_annotations[transcript_id]['num_ejcs_remaining'] = num_ejcs_remaining
    ejc_annotations[transcript_id]['interval'] = stop_codon

ejc_annotations = pd.DataFrame(ejc_annotations).transpose()

#converts information from transcript level to gene level
def special_dict():
    return {"distance_from_three_prime_ends" : [], "intervals" : [], "num_trailing_exons" : []} 
gene_annotations = defaultdict(special_dict)

for index, series in ejc_annotations.iterrows():
    gene_annotations[transcript_gene_dict[index]]["distance_from_three_prime_ends"].append(series['distance_from_three_prime_end'])
    gene_annotations[transcript_gene_dict[index]]["intervals"].append(series['interval'])
    gene_annotations[transcript_gene_dict[index]]["num_trailing_exons"].append(series['num_ejcs_remaining'])
    
gene_annotations = pd.DataFrame(gene_annotations).transpose()

#adds nmd annotations based off of litetrature information
nmd_counts = []
for index, gene in gene_annotations.iterrows():
    nmd_count = 0
    for distance_from_three_prime_end, num_ejc in zip(gene['distance_from_three_prime_ends'], gene['num_trailing_exons']):
        if distance_from_three_prime_end < -50 and num_ejc >= 1:
            nmd_count += 1
    nmd_counts.append(nmd_count)
gene_annotations['nmd_transcripts'] = nmd_counts
```

```python
five_prime_utrs = gff.filter(lambda x: x[2] == "five_prime_UTR").filter(lambda x: x.chrom in (map(str, range(1,23)) + ['X', 'Y'])).each(make_ucsc_chr).saveas()
five_prime_utrs.sequence(fi="/nas3/yeolab/Genome/ucsc/hg19/chromosomes/all.fa", s=True, fo="five_prime_utr.fasta")

sequences = SeqIO.parse(open("five_prime_utr.fasta"), 'fasta')
five_prime_utr_dict = defaultdict(list)

for five_prime_utr, sequence in zip(five_prime_utrs, sequences):
    five_prime_utr_dict[five_prime_utr.attrs['Parent']].append((five_prime_utr, sequence))

for five_prime_utr in five_prime_utr_dict.values():
    if five_prime_utr[0][0].strand == "-":
        five_prime_utr.reverse()
```

```python
uorf_annotations = ""
#By visual examination everything is correct, need to check in detail before publishing...
#should check number of annotations to see how well they match up with other reports...
for five_prine_utr in five_prime_utr_dict.values():
    for reading_frame in range(3):
        first = True
        in_uORF = False
        offset = ""
        offset_size = 0
        for interval, sequence in five_prine_utr:
            #does offsetting for all reading frames, but scanning starts at the first exon, so only offset that one
            if first:
                exon = sequence.seq[reading_frame:]
                first = False
            else:
                exon = sequence.seq
                
            for x, amino_acid in enumerate((offset + exon).translate()):
                if interval.strand == "+":
                    start_of_codon = interval.start + offset_size + reading_frame + (x * 3)
                    end_of_codon = interval.start + offset_size + reading_frame + ((x + 1) * 3)
                else:
                    end_of_codon = interval.end - offset_size - reading_frame - (x * 3) 
                    start_of_codon = interval.end - offset_size  - reading_frame - ((x + 1) * 3)
                    
                if amino_acid == "M" and not in_uORF:
                    start_annotation = "\t".join([interval.chrom, "protein_coding", "uORF_start", str(start_of_codon), str(end_of_codon), ".", interval.strand, ".", "ID=uORF_start:%s;Parent=%s" % (interval.attrs['Parent'], interval.attrs['Parent'])])
                    uorf_start = x
                    in_uORF = True
                
                if amino_acid == "*" and in_uORF:
                    in_uORF = False 
                    stop_annotation = "\t".join([interval.chrom, "protein_coding", "uORF_end", str(start_of_codon), str(end_of_codon), ".", interval.strand, ".", "ID=uORF_end:%s;Parent=%s" % (interval.attrs['Parent'], interval.attrs['Parent'])])

                    if x - uorf_start > 30:
                        uorf_annotations += start_annotation + "\n"
                        uorf_annotations += stop_annotation + "\n"
     
            offset_size = len(sequence.seq) % 3
            offset = sequence.seq[offset_size * -1:]
            
uorf_annotations = pybedtools.BedTool(uorf_annotations, from_string=True).saveas()
        
```

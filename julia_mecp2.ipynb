#Code For Loading RPKMs

```python
datasets = glob.glob("/nas3/gpratt/projects/julia/analysis/mecp2_ripseq_v1/*.rpkm")
datasets = {os.path.basename(dataset).split("_")[0] : pd.read_csv(dataset, index_col=0, sep="\t") for dataset in datasets}
ripseq_df = pd.concat(datasets, join="outer")
ripseq_df = ripseq_df.drop("flag",1)
ripseq_df['gene_id'] = ripseq_df.index.get_level_values(1)
ripseq_df['gene_name'] = ripseq_df.gene_id.apply(lambda x: gene_name_dict[x])
ripseq_df['gene_type'] = ripseq_df.gene_id.apply(lambda x: gene_type_dict[x])
ripseq_df = ripseq_df.drop("gene_id",1)
```

```python
mecp2_combined = ripseq_df.ix['MeCP2-IP'] 
mecp2_input = ripseq_df.ix['MeCP2-input']

mecp2_combined["Log2 IP / Input Fold Change"] = (mecp2_combined.RPKM / mecp2_input.RPKM).apply(log2)

mecp2_combined = mecp2_combined.drop("RPKM", 1)
mecp2_combined = mecp2_combined.drop("logged_rpkm", 1)
mecp2_combined['ip_rpkm'] = ripseq_df.ix['MeCP2-IP'].RPKM
mecp2_combined['input_rpkm'] = ripseq_df.ix['MeCP2-input'].RPKM
mecp2_combined.sort("ip_rpkm", ascending=False)
```

That didn't look interesting, what happens if I sort by fold change, and because there is low coverage and some of the input genes are covered, I'll do a second sort based on expression in the IP sample

```python
covered_ips = mecp2_combined[~(mecp2_combined['ip_rpkm'] == 0)] 
covered_ips.sort(["Log2 IP / Input Fold Change", 'ip_rpkm'], ascending=False)
```

That didn't turn up anything interesting, lots of antisense and pseudogenes.  Next looking at only protein coding genes

```python
covered_ips = mecp2_combined[~(mecp2_combined['ip_rpkm'] == 0) & (mecp2_combined['gene_type'] == 'protein_coding')] 
covered_ips.sort(["Log2 IP / Input Fold Change", 'ip_rpkm'], ascending=False)
```

Look just at protein coding genes that have 2 fold or greater change 

```python
mecp2_fold_filter = mecp2_combined[~isinf(mecp2_combined["Log2 IP / Input Fold Change"])].dropna().sort("Log2 IP / Input Fold Change", ascending=False)
mecp2_fold_filter = mecp2_fold_filter[(mecp2_combined['gene_type'] == 'protein_coding') & (mecp2_fold_filter["Log2 IP / Input Fold Change"] > 2)]
```

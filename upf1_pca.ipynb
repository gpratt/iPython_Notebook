```python
from sklearn import decomposition
import pandas
import numpy as np
import scipy as sp
import pylab as pl
from collections import defaultdict
from glob import glob
import itertools
```

```python
df = pandas.read_csv('/nas3/gpratt/projects/upf1/analysis/rna_seq/all_rpkms/all.txt', sep='\t', index_col=0)
print shape(df)

df_t = df.transpose()
print shape(df_t)

colors_array = []
names_array = []
for column in df:

    if column == 'gene_ID':
        pass
    
    else:
        
        names_array.append(str('${}$'.format(column)).replace('_', '-').replace('.gz.bam.sorted.bam.rpkm', ''))
        if column[0] == '3':
            colors_array.append('r')
        elif column[0] == 'c':
            colors_array.append('g')
            
print names_array
```

```python
df_array = df_t.as_matrix()
#values = df_array
values = log(df_array)
print shape(values)

pca = decomposition.PCA(whiten=True, n_components=3)
pca.fit(values)
X=pca.transform(values)
```

```python
gene_ids = df.index
print gene_ids
```

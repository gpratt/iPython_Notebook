```python
from IPython.core.display import HTML

from gscripts.general import parsers
```

```python
qc_metrics = parsers.clipseq_metrics("/nas3/gpratt/projects/tiffany_genentech/analysis/v1/", iclip=True)
qc_metrics = qc_metrics.rename(columns={"Processed bases" : "Input Bases",
           "Processed reads" : "Input Reads",
           "Number of input reads" : "Reads after Quality Filtering"
           })
```

#Raw QC

```python
HTML(qc_metrics.to_html())
```

#Important QC Metrics

```python
qc_metrics["Reads After Triming"] = qc_metrics['Input Reads'] - qc_metrics['Too short reads']

filtered_df = qc_metrics[[ "Input Reads",
                            "Reads After Triming",
                            "Reads after Quality Filtering", 
                            "Uniquely mapped reads number",
                            #"Uniquely mapped reads %",
                            "Usable Reads",
                            #"spot",
                            #"Num Peaks"
                            ]]

filtered_df['Percent Usable'] = (filtered_df['Usable Reads'] / filtered_df['Uniquely mapped reads number']) * 100
HTML(filtered_df.to_html(formatters={"Input Reads" : parsers.commas,
                                     "Reads After Triming" : parsers.commas,
                                     "Reads after Quality Filtering" : parsers.commas,
                                     "Uniquely mapped reads number" : parsers.commas,
                                     "Usable Reads" : parsers.commas,
                                     "Num Peaks": parsers.commas
                                     } ))
```

Analysis
--------

The number of usable (randomer collapsed) reads from this analysis is too low to confidently draw genome wide conclusions.  The Precent Usable Reads is decent when compared to internal datasets from the Yeo Lab.  We see a range of 1%-10% on average.   

I recommend much deeper sequencing.  Enough to get up to ~1m usable reads.  For R08 this means sequencing at least 23,000,000 reads and 17,000,000 for R02

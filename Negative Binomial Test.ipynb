Estimating negative binomial... need to estimate r and p

display(Math(p = \frac{ \sum_{i=1}^N k_i / N } {r + \sum_{i=1}^N k_i / N }))

Equaltions got from here:  http://en.wikipedia.org/wiki/Negative_binomial_distribution

```python
p_estimate = lambda total_result, r: (sum(total_result) / len(total_result)) / (r + (sum(total_result) / len(total_result))) 
log_likelihood_negative_binomial = lambda total_result, r, p: sum(log(gamma(total_result + r))) - sum(log(scipy.misc.factorial(total_result))) - len(total_result) * log(gamma(r)) + len(total_result) * r * log(1-p) + sum(total_result * log(p))
partial_wrt_r = lambda total_result, r: sum(scipy.special.digamma(total_result + r)) - sum(total_result) * scipy.special.digamma(r) + sum(total_result) * log(r / (r + (sum(total_result) / len(total_result))))
```

```python
prev_r = -9999
cur_r  = 5

while abs(prev_r - cur_r) > .05:
    print "running"
    prev_r = cur_r
    print "cur_r", cur_r
    p = p_estimate(total_result, cur_r)
    print "p", p
    cur_r = max(cur_r - log_likelihood_negative_binomial(total_result, cur_r, p) / partial_wrt_r(total_result, cur_r),0)
    
```

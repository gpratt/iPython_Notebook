```python
def p_estimate(k, r):   
    return np.mean(k) / (r + np.mean(k)) 
```

```python
def log_likelihood_negative_binomial(k, r, p):
    return (sum(log(gamma(k + r))) - 
           sum(log(factorial(k))) - 
           len(k) * log(gamma(r)) + 
           (len(k) * r * log(1-p)) + 
           sum(k * log(p)))
```

```python
def partial_wrt_r(k, r):
    return (sum(digamma(k + r)) - 
            len(k) * digamma(r) + 
            len(k) * log(r / (r + np.mean(k))))
```

```python
def is_valid_bed12(interval):
    
    """
    
    Removes clusters that start in invalid locations for bed12 formatted files
    
    x - bedline
    
    returns either true if line is valid 
    
    """
    
    tstart = int(interval[6])
    tstop = int(interval[7])
    return interval.start < tstart and interval.stop > tstop
```

#See if there is a correlation between peaks with high expression and p-value
Will also need to seperate by containing and not containing motif 

```python
from bx.bbi.bigwig_file import BigWigFile
import pandas as pd
```

```python
neg_strand = BigWigFile(open("/Users/gpratt/bioinformatics/yeolab/projects/fox2_iclip/analysis/fox2_iclip_rnaseq_v1/SRR988462.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.neg.t.bw"))
pos_strand = BigWigFile(open("/Users/gpratt/bioinformatics/yeolab/projects/fox2_iclip/analysis/fox2_iclip_rnaseq_v1/SRR988462.fastq.gz.polyATrim.adapterTrim.rmRep.sorted.rg.pos.bw"))
peaks = pybedtools.BedTool("/Users/gpratt/bioinformatics/yeolab/projects/fox2_iclip/analysis/ad-hoc/FOX2.em.peaks.bed")
gcaug_motif = pybedtools.BedTool("/Users/gpratt/bioinformatics/yeolab/projects/fox2_iclip/analysis/ad-hoc/motif_GCATG.BED")
overlapping_peaks = peaks.intersect(gcaug_motif, s=True, u=True).to_dataframe()
```

```python
p_value = []
mean_coverage = []
for interval in peaks:
    if interval.strand == "+":
        cur_coverage = neg_strand.get_as_array(interval.chrom, interval.start, interval.end)
    elif interval.strand == "-":
        cur_coverage = pos_strand.get_as_array(interval.chrom, interval.start, interval.end)
    
    cur_coverage = np.mean([value for value in np.abs(cur_coverage) if not np.isnan(value)])
    mean_coverage.append(cur_coverage)
```

```python
peaks_df = peaks.to_dataframe()
peaks_df['mean_coverage'] = mean_coverage
peaks_df['log10_p_value'] = -1 * np.log10(peaks_df.score)
peaks_df['has_gcaug'] = [item in overlapping_peaks.name.values for item in peaks_df.name]
peaks_df = peaks_df.dropna()
```

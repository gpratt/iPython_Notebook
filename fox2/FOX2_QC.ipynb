```python
def get_peak_wiggle(tool, data_pos, data_neg, size = 1000, num_peaks=100):
    
    """
    
    returns matrix of (n,25) of wiggle and peak information
    tool - bedtool contains location of peaks
    data - bigwig file handle - contains bw data for the peaks

    """
    
    results = []

    for line in tool[:num_peaks]:
        peak_center = (int(line[6]) + int(line[7])) / 2
        pad_amt = size / 2
        if line.strand == "+":
            result = data_pos.get_as_array(line.chrom, peak_center - pad_amt , peak_center + pad_amt)
        if line.strand == "-":
            result = data_neg.get_as_array(line.chrom, peak_center - pad_amt , peak_center + pad_amt)[::-1]
            
        result = [0 if np.isnan(x) else x for x in result]
        

        #print len(result)
        results.append(result)
        #print len(result)

    return pd.DataFrame(results)
```

```python
import math

def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
```

```python
def AR(A,B):
    result = 0 
    """ A, B are clusters, dicts with indicides 0-n that contain lists of locations that belong to that cluster"""
    t1 = sum(nCr(len(x), 2) for x in A.values())
    t1 = sum(nCr(len(x), 2) for x in B.values())
    t3 = (2 * t1 * t2) / sum(len(x) for x in A.values()) * (sum(len(x) for x in A.values()) - 1)
    for x in A.values():
        for y in B.values():
             result += len(x | y) - t3
                
    return result / .5 * (t1 * t2) - t3
```
